## Основные теоретические положения
Случайный вектор $\vec{\xi} = (\xi_1, ..., \xi_n)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1, ..., \eta_m)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\vec{B} = (b_1, ..., b_n)^\intercal$, такие что: 
  $\vec{\xi} = A\vec{\eta} + \vec{b}$

Плотность распределения нормального случайного вектора имеет вид:

$p_{\vec{\xi}}(x_1, ..., x_n) = \frac1{(2\pi)^{\frac{n}2} \sqrt{\det{\Sigma}}} \exp\left(-\frac12 (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \right)$,

где $ \vec{x}=(x_1,...,x_n)^{\intercal} $

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределённой.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределёнными.
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}\left(\vec{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, а B - произвольный вектор размерности $n$, то $A\vec{\xi} + B \sim \mathcal{N}\left( A\vec{\mu}+B, A \Sigma A^\intercal \right)$.

Условное распределение компоненты нормального случайного вектора:
$p_{\xi|\eta=y}(x) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$,
где $p_\eta(y)$ - одномерное нормальное распределение:
$ p_{\eta}(y) = \frac{1}{\sqrt{2\pi}\sqrt{\mathbb{D}_{\eta}}} \exp\left(\frac{-\frac{1}{2}(y-\mathbb{E}_{\eta})^2}{\mathbb{D}_{\eta}}\right) $

#### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi,\eta}(x,y)$.
$p_{\xi,\eta}(x,y)=C\cdot\exp\left(-\frac{1}{2}(3x^2-3xy+7y^2-9x+17y+13)\right)$ <br>
$q(x,y)=(3x^2-3xy+7y^2-9x+17y+13)=3(x^2-xy-3x)+7y^2+17y+13 = $ <br>
$= 3(x^2-x(y+3)+\frac{(y+3)}{4})^2+7y^2+17y+13-\frac{3(y+3)^2}{4} = $ <br>
$= 3(x^2-x(y+3)+\frac{(y+3)}{4})^2+7y^2+17y+13-\frac{3y^2+18y+27}{4} = $ <br>
$= 3(x^2-x(y+3)+\frac{(y+3)}{4})^2+7y^2+17y+13-\frac{3y^2}{4}- \frac{18y}{4} - \frac{27}{4}= $ <br> 
$= 3(x^2-x(y+3)+\frac{(y+3)}{4})^2+6\cdot\frac{1}{4}(y+1)^2 = $ <br>
$= 3((x+1)+\frac{1}{2}(y+1))^2+6\cdot\frac{1}{4}(y+1)^2)=$ <br>
$= 3(x+1)^2+3(x+1)(y+1)+\frac{3}{4}(y+1)^2+6\cdot\frac{1}{4}(y+1)^2 =$ <br>
$= 3(x+1)^2+3(x+1)(y+1)+7(y+1)^2$ <br>
$\Sigma^{-1}= \begin{pmatrix}3&\frac32\\\frac32&7\end{pmatrix}$
$ E = \begin{pmatrix}\xi\\\eta\end{pmatrix}=\begin{pmatrix}-1\\-1\end{pmatrix} \\ $ 
$ \Sigma=\frac7{75}\begin{pmatrix}7&-\frac32\\-\frac32&3\end{pmatrix}\\$
$21-\frac94=18\frac34=\frac{75}4=det\Sigma^{-1}$
$D_\xi=\frac{28}{25};\;\;D_\eta=\frac{12}{75}=\frac4{25} \\$
$ cov(\xi,\eta)=-\frac{12}{180}=-\frac2{25} \\$
$C=\frac1{2\mathrm\pi\cdot\sqrt{\displaystyle\frac4{75}}}=\frac1{2\mathrm\pi\cdot{\displaystyle\frac2{5\sqrt3}}}=\frac{5\sqrt3}4 \\$
$\rho(\xi,\eta)=\sqrt{\frac{75}{28}}\;\cdot\;\sqrt{\frac{25}4}\;\cdot\;\left(-\frac2{25}\right)\;=\;\frac{3\sqrt5}{2\sqrt7}\cdot\;\frac52\cdot\;\left(-\frac2{25}\right)=\;-\frac3{10} \\$





```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
mu = np.array([-1, 1]) # вектор мат ожидания
cov = np.array([[28/25, -2/25], [-2/25, 4/25]]) #  матрица ковариации
inv_cov = np.linalg.inv(cov) # обратная ковариационная матрица
det_cov = np.linalg.det(cov) # det  матрицы ковариации

def pdf(x, y):
    xy = np.array([x, y]) - mu
    exponent = -0.5 * np.dot(np.dot(xy, inv_cov), xy)
    return (1 / (2 * np.pi * np.sqrt(det_cov))) * np.exp(exponent)

Z = np.vectorize(pdf)(X, Y)
#график
grf = plt.figure()
A = grf.add_subplot(111, projection='3d')
A.plot_surface(X, Y, Z, cmap='viridis')
A.set_xlabel('x')
A.set_ylabel('y')
plt.show()
```
