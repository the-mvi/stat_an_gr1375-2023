## Основные теоретические положения

Случайный вектор $\vec{\xi}=(\xi_{1},\ldots,\xi_{n})^\intercal \colon \Omega \to \mathbb{R}^n $ называют <b><i> нормальным </i></b>, если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta}= (\eta_{1},\ldots,\eta_{n})^\intercal$, матрица $A$ размерности $m \times n$ и вещественный вектор $B=(b_{1},\ldots,b_{n})^\intercal$, такие что:  $\vec{\xi}=A\vec{\eta}+B$ <br>
<i>Обозначение</i>:  $\vec{\xi} \sim \mathcal{N}(\mu, \Sigma)$ где $\mu = \mathbb{E}\vec{\xi}=(\mathbb{E}\xi_1,\ldots,\mathbb{E}\xi_n)^\intercal$ - вектор математических ожиданий компонент; $\Sigma = \{{\sigma_{i,j}}\}_{i,j=1}^n$ ; $\sigma_{i,j}=\mathop{\mathrm{cov}}\nolimits(\xi_i,\xi_i)$ - ковариационная матрица<br>
<b><i>Плотность нормального вектора</i></b> в терминах матрицы имеет вид:<br>
$p_{\vec{\xi}}(x_{1},\ldots,x_{n})= \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{det\Sigma}} \exp\left(-\frac{1}{2}(\vec{x}-\mu)^\intercal \cdot \Sigma^{-1} \cdot (\vec{x}-\mu)\right)$, где $\vec{x}=(x_{1},\ldots,x_{n})^\intercal$ <br>
<b><i>Свойства многомерного нормального распределения </i></b>: <br>
<ul style="list-style-type:square">
 <li>Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными.</li>
 <li>Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями.</li>
 <li>Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.</li>
 <li>Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}(\mu, \Sigma)$, 
 $A$- произвольная матрица неособенная размерности $m \times n$ , а $B$ - произвольный вектор размерности $m \times 1$ , то <br>  $ A\vec{\xi}+B \sim \mathcal{N}(A\mu+B, A \Sigma A^\intercal) $  </li>
</ul>

<b><i>Ковариационная матрица: </i></b><br>
$ \Sigma = \begin{pmatrix} 
   \mathop{\mathrm{cov}}\nolimits(\xi,\xi) & \mathop{\mathrm{cov}}\nolimits(\xi,\eta) \\
   \mathop{\mathrm{cov}}\nolimits(\eta,\xi) & \mathop{\mathrm{cov}}\nolimits(\eta,\eta)
  \end{pmatrix} 
$ <br>
$\mathop{\mathrm{cov}}\nolimits(\xi,\xi) = \mathbb{D}\xi$ <br>
$\mathop{\mathrm{cov}}\nolimits(\eta,\eta) = \mathbb{D}\eta$ <br>
<b><i>Коэффициент корреляции: </i></b><br>
 $\rho(x,y)=\frac{\mathop{\mathrm{cov}}\nolimits(x,y)}{\sqrt{\mathbb{D}(x)\mathbb{D}(y)}}$<br>
 <b><i> Аффинное преобразование  </i></b> $f\colon\mathbb{R}^n\to \mathbb{R}^n$ есть преобразование вида: <br>
$f(x)=M \cdot x + v$, <br>
где $M$ - обратимая матрица и $v \in \mathbb{R}^n$ <br>
 <b><i>Условной плотностью </i></b>$p(x|y)$ распределения составляющей $X$ при $Y=y$ называется отношение плотности $p(x,y)$ двумерной случайной величины $(X,Y)$ к плотности распределения $p(y)$ при условии, что составляющая $Y$ приняла конкретное значение или попала в заданный интервал. То есть <br>
$p(x|y)=\frac{p(x,y)}{p(y)}$ <br>
