## Основные теоретические положения
#### Определение нормального вектора
 Случайный вектор $\vec{\xi}=\left(\xi_1,\ldots,\xi_n\right)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta}=\left(\eta_1,\ldots,\eta_n \right)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\vec{b} = \left(b_1,\ldots,b_n \right)^\intercal$, такие что $\vec{\xi} = A\vec{\eta} + \vec{b}$ <br>
 
 Обозначение: $\vec{\xi} \sim \mathcal{N} \left(\mu, \Sigma \right)$ <br>
   где $ \mu = \mathbb{E}\vec{\xi}=\left(\mathbb\xi_1,\ldots,\mathbb\xi_n \right)^{\intercal}$ - вектор математических ожиданий компонент; <br>
   
 $\Sigma = \{ \mathrm{cov} \left(\xi_i, \xi_j \right) \}_{i,j=1,1}^n $ - ковариационная матрица. <br>
 
 Плотность распределения нормального случайного вектора имеет вид: <br>
 
   $p_{\vec{\xi}}\left(x_1,\ldots, x_n \right) = \frac1{\left(2\pi \right)^{\frac{n}{2}} \sqrt{\det{\Sigma}}} \exp\left(-\frac{1}{2} (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \right)$ <br>
   
   Плотность условного распределения для $\xi$: $ p(\xi|\eta=y) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$
#### Многомерное нормальное распределение
*Многомерное нормальное распределение является предельным для распределений нормированных сумм случайных векторов. В частности, распределение (мультиномиальное) вектора нормированных частот*$\left( n_{1}^*,\ldots,n_{N-1}^* \right)=\vec{n_{}^*},n_{i}^*=\left(\upsilon_{i}-np_{i}\right)/\sqrt{n}$ *стремиться к вырожденному нормальному распределению*$N\left(0,\Sigma=\lVert \sigma_{i,j} \rVert_{1}^{N-1}\right)$, где $\sigma_{i,j}=p_{i}\left(1-p_{i}\right),\sigma_{i,j}=p_{i}p_{j},i \neq j$ *(ковариации мульномиального распределения). Известно также, что квадратичная форма* $Y^{T}\Sigma^{-1}Y$ *, стоящая в показателе экспоненты, показывающей плотность невырожденного нормального распределения* $N\left(0,\Sigma\right)$ *, распределена по закону* $\mathcal{X}_(N-1)^2.^1$ *.Данное свойство лежит в основе* $\mathcal{X}^2$ *критерия*

 Аффинное преобразование $ f:\mathbb{R}^n \rightarrow \mathbb{R}^n$ есть преобразование вида:
   $ f \left(x\right)=M \cdot x + \upsilon,$
   где M - обратимая матрица $\upsilon \in \mathbb{R}^n$

#### Задание 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.
$q\left(x,y\right)=3x^2-6xy+11y^2+16y+8=\left(3x^2-6xy\right)+11y^2+16y+8=3\left(x^2+2xy\right)+11y^2+16y+8=3\left(x^2-2xy+y^2\right)-3y^2+11y^2+16y+8=3\left(x-y\right)^2+8\left(y+1\right)^2=3\left(x-\left(y+1\right)+1\right)^2+8\left(y+1\right)^2=3\left(\left(x+1\right)^2-2\left(x+1\right)\left(y+1\right)+\left(y+1\right)^2\right)+8\left(y+1\right)^2=3\left(x+1\right)^2-2*3\left(x+1\right)\left(y+1\right)+11\left(y+1\right)^2$
##### Вектор мат. ожидания:
 $ \mathbb{E}  \begin{pmatrix} \xi \\ \eta \end {pmatrix} = \begin{pmatrix} -1 \\ -1 \end{pmatrix} $ 
##### Ковариационная матрица:
 $\Sigma^{-1} = \begin{pmatrix} 3 & -3 \\ -3 & 11 \end{pmatrix} \Rightarrow \Sigma = \frac{1}{24}\begin{pmatrix} 11 & 3 \\ 3 & 3 \end{pmatrix}$ 
##### Ковариационные характеристики:
 $D\left(\xi\right)=\frac{1}{8}  ;$
 $D\left(\eta\right)=\frac{11}{24}  ;$
 $cov\left(\xi,\eta\right)=-\frac{1}{8}  ;$ - ковариация
 $\rho=-\frac{3}{24\sqrt{\frac{1*11}{8*24}}}=-\frac{3}{\sqrt{33}}  ;$ - кореляция
 $  C=\frac{1}{2\pi^{\frac{n}{2}}\sqrt{det\xi}}=\frac{1*\sqrt{24}}{2\pi*1}=\frac{\sqrt{6}}{\pi}  $
##### Плотность нормального распределения:
 $P_{\xi,\eta}\left(x,y\right)=\frac{\sqrt{6}}{\pi}exp\left\lbrace -\frac{1}{2} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} -1 \\ -1 \end{pmatrix} \right\rbrack^{\intercal} \begin{pmatrix} 3 & -3 \\ -3 & 11 \end{pmatrix} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} -1 \\ -1 \end{pmatrix} \right\rbrack \right\rbrace$

```python
import plotly.graph_objects as go
import numpy as np

def func(X,Y):
    Z = (np.sqrt(6)/(np.pi))*np.exp(-1/2*(3*(X+1)**2-6*(X+1)*(Y+1)+11*(Y+1)**2))
    return Z
x =  np.linspace(-4, 3, 100)
y =  np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x,y)
Z = func(X, Y)

fig = go.Figure(data=go.Surface(x=X, y=Y, z=Z, colorscale="Spectral"))

fig.update_layout(
    title="График плотности распределения",
    scene=dict(
        xaxis=dict(title="X"),
        yaxis=dict(title="Y"),
        zaxis=dict(title="Z")
    )
)
fig.show()
```

#### Задание 3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор с независимыми компонентами.
##### Собственные числа и векторы ковариационной матрицы
$det\left(\sum^{-1}\lambda E\right)$ <br>
 $\begin{pmatrix} 3-\lambda & -3 \\ -3 & 11-\lambda \end{pmatrix}=\left(3-\lambda\right)\left(11-\lambda\right)-9=33-3\lambda-11\lambda+\lambda^2-9=\lambda^2-14\lambda+24$ <br>
   $\lambda_{1}=2;$ $\lambda_{2}=12$ <br>
  $\lambda_{1}=2:\begin{pmatrix} 1 & -3 \\ -3 & 9 \end{pmatrix}\Rightarrow\begin{pmatrix} 1 & -3 \end{pmatrix}\Rightarrow x-3y=0 \Rightarrow \vec{\vartheta}'=\frac{1}{\sqrt{10}}\begin{pmatrix} 3 \\ 1 \end{pmatrix}$ <br>
 $\lambda_{2}=12:\begin{pmatrix} -9 & -3 \\ -3 & -1 \end{pmatrix}\Rightarrow\begin{pmatrix} -3 & -1 \end{pmatrix}\Rightarrow -3x-y=0 \Rightarrow \vec{\vartheta_{2}}'=\frac{1}{\sqrt{10}}\begin{pmatrix} 1 \\ -3 \end{pmatrix}$ <br>
 $Q^{\intercal}=\frac{1}{\sqrt{10}}\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}$
##### Случайный вектор с независимыми компонентами
 $\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}=\frac{1}{\sqrt{10}}\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix}$
##### Вектор математического ожидания
 $E\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}=\frac{1}{\sqrt{10}}\begin{pmatrix} 3 & 1 \\ 1 & -3\end{pmatrix} \begin{pmatrix} -1 \\ -1 \end{pmatrix}=\frac{1}{\sqrt{10}} \begin{pmatrix} -4 \\ 2 \end{pmatrix}$
##### Ковариационная матрица
 $\sum_{3}=\frac{1}{24}*\frac{1}{10}\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}\begin{pmatrix} 11 & 3 \\ 3 & 3 \end{pmatrix}\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}=\frac{1}{240}\begin{pmatrix} 120 & 0 \\ 0 & 20 \end{pmatrix}$ <br>
 $\sum_{3}^{-1}=\begin{pmatrix} 2 & 0 \\ 0 & 12 \end{pmatrix}$
##### Плотность нормального распределения
 $C=\frac{1}{2\pi\frac{1}{\sqrt{24}}}=\frac{\sqrt{6}}{\pi}$ <br>
 $P_{\xi,\eta}=\frac{\sqrt{6}}{\pi}exp\left\lbrace-\frac{1}{2}\left(2\left(x+\frac{4}{\sqrt{10}}\right)^2\right)+12\left(y-\frac{2}{\sqrt{10}}\right)^2 \right\rbrace$

```python
import plotly.graph_objects as go
import numpy as np

def func(X,Y):
    Z = (np.sqrt(6)/(np.pi))*np.exp( (-1/2)* (2*((X+4)/np.sqrt(10))**2 + 12*((Y-4)/np.sqrt(10))**2) )
    return Z
x =  np.linspace(-13, 4, 100)
y =  np.linspace(0, 10, 100)
X, Y = np.meshgrid(x,y)
Z = func(X, Y)

fig = go.Figure(data=go.Surface(x=X, y=Y, z=Z, colorscale="Spectral"))

fig.update_layout(
    title="График плотности распределения",
    scene=dict(
        xaxis=dict(title="X"),
        yaxis=dict(title="Y"),
        zaxis=dict(title="Z")
    )
)
fig.show()
```

#### Задание 4. Вычислить характеристики совместного распределения случайного вектора $\left(-2\xi - 3\eta,-2\xi - 4\eta \right)^{\intercal} $ и записать его плотность.
 $\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}=\begin{pmatrix} -2\xi - 3\eta \\ -2\xi - 4\eta \end{pmatrix}=\begin{pmatrix} -2 & -3 \\ -2 & -4 \end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$ <br>
 
 $E\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}=\begin{pmatrix} -2 & -3 \\ -2 & -4 \end{pmatrix}\begin{pmatrix} -1 \\ -1 \end{pmatrix}=\begin{pmatrix} 4 \\ 2 \end{pmatrix}$ <br>
 
 $\sum_{4}=\frac{1}{24}\begin{pmatrix} -2 & -3 \\ -2 & -4 \end{pmatrix}\begin{pmatrix} 11 & 3 \\ 3 & 3 \end{pmatrix}\begin{pmatrix} -2 & -2 \\ -3 & -4 \end{pmatrix}=\frac{1}{24}\begin{pmatrix} 107 & 122 \\ 122 & 140 \end{pmatrix}$ <br>
 
 $detZ_{4}=\frac{1}{6};$ $D_{\xi_{4}}=\frac{107}{24};$ $D_{\eta_{4}}=\frac{140}{24};$ $cov\left(\xi_{4},\eta_(4)\right)=\frac{122}{24};$ $\rho\left(\xi_{4},\eta_{4}\right)=\frac{cov\left(\xi_{4},\eta_{4}\right)}{\sqrt{D_{\xi_{4}}}*D_{\eta_{4}}}=\frac{122}{24\frac{\sqrt{107*140}}{24}}=\frac{61}{\sqrt{3745}}$ <br>
 
 $\sum_{4}^{-1}=6\begin{pmatrix} \frac{140}{24} & -\frac{122}{24} \\ -\frac{122}{24} & \frac{107}{24} \end{pmatrix}=\begin{pmatrix} \frac{140}{4} & -\frac{122}{4} \\ -\frac{122}{4} & \frac{107}{4} \end{pmatrix}$ <br>
 ##### Плотность совместного распределения случайного вектора
 $C=\frac{1}{2\pi\sqrt{det\sum_{4}}}=\frac{\sqrt{6}}{2\pi}$ <br>
 
 $P_{\xi_{4}\eta_{4}}\left(x,y\right)=\frac{\sqrt{6}}{2\pi}exp\left\lbrace -\frac{1}{2} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} 5 \\ 6 \end{pmatrix} \right\rbrack^{\intercal} \begin{pmatrix} \frac{140}{4} & -\frac{122}{4} \\ -\frac{122}{4} & \frac{107}{4} \end{pmatrix} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} 5 \\ 6 \end{pmatrix} \right\rbrack \right\rbrace$

```python
import plotly.graph_objects as go
import numpy as np

def func(X,Y):
    Z = (np.sqrt(6)/(2*np.pi))*np.exp(-1/2 * ((140/4) * (X-5)**2 - 2 *(122/4)* (X-5) * (Y-6) + (107/4)*(Y-6)**2))
    return Z
x =  np.linspace(-5, 15, 50)
y =  np.linspace(-4, 16, 50)
X, Y = np.meshgrid(x,y)
Z = func(X, Y)

fig = go.Figure(data=go.Surface(x=X, y=Y, z=Z, colorscale="Spectral"))

fig.update_layout(
    title="График плотности распределения",
    scene=dict(
        xaxis=dict(title="X"),
        yaxis=dict(title="Y"),
        zaxis=dict(title="Z")
    )
)
fig.show()
```
