## Основные теоретические положения
Двумерное нормальное распределение - это вероятностное распределение двух случайных переменных, которые коррелированы и имеют нормальное распределение. Основные положения двумерного нормального распределения:

1. Математическое ожидание: Математическое ожидание двумерного нормального распределения задается вектором средних значений, обозначаемым как $\mathbb{E}$. Он представляет собой центр распределения и определяет среднее значение каждой случайной переменной.

2. Ковариационная матрица: Ковариационная матрица двумерного нормального распределения, обозначаемая как $\Sigma$, определяет степень корреляции и дисперсию каждой переменной.

3. Плотность распределения: Плотность вероятности двумерного нормального распределения задается формулой:
$f(x,y) = \frac{1}{2\pi \cdot \sigma_{x} \cdot \sigma_{y} \cdot \sqrt{1 - \rho^{2}}} \cdot exp^{-\frac{1}{2 \cdot(1-\rho^{2})} \cdot \left[ \frac{(x - \mathbb{E}_{x})^{2}}{\sigma_{x}^{2}} - \frac{2\rho \cdot(x-\mathbb{E}_{x})\cdot(y-\mathbb{E}_{y})}{\sigma_{x} \cdot \sigma_{y}} + \frac{(y-\mathbb{E}_{y})^{2}}{\sigma_{y}^{2}} \right]}$

4. Форма распределения: Двумерное нормальное распределение имеет форму эллипса в двумерном пространстве. Ориентация и размер эллипса определяются ковариационной матрицей, а среднее значение задает центр эллипса.

5. Корреляция: Коэффициент корреляции $\rho$ определяет степень линейной связи между переменными $x$ и $y$. Значение $\rho$ находится в диапазоне от -1 до 1, где отрицательное значение указывает на обратную связь, положительное - на прямую связь, а нулевое - на отсутствие линейной связи.

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $P_{\xi,\eta}$

Для выполнения задания начнем с преобразования Лагранжа: 

$q(x,y) = 6x^{2}+4xy+9y^{2}-8x+14y+11 = (6x^{2}+4x(y-2))+9y^{2}+14y+11 = 6\left(x^{2}+\frac{2}{3}x(y-2)\right)+9y^{2}+14y+11 = 6\left(x^{2}+\frac{2}{3}x(y-2)+\left(\frac{y-2}{3}\right)^{2}\right)-6\left(\frac{y-2}{3}\right)^{2}+9y^{2}+14y+11 = 6\left(x+\frac{y-2}{3}\right)^{2}-6\frac{y^{2}-4y+4}{9}+9y^{2}+14y+11 = 6\left(x+\frac{4}{3}-\frac{2}{3}\right)^{2}-\frac{2y^{2}-8y+8}{3}+9y^{2}+14y+11 = 6\left(x+\frac{4}{3}-\frac{2}{3}\right)^{2}-\frac{2y^{2}}{3}+\frac{8y}{3}-\frac{8}{3}+9y^{2}+14y+11 = 6\left(x+\frac{4}{3}-\frac{2}{3}\right)^{2}-\frac{2y^{2}}{3}+\frac{27}{3}y^{2}+\frac{8y}{3}+\frac{42}{3}y-\frac{8}{3}+\frac{33}{3} = 6\left(x+\frac{4}{3}-\frac{2}{3}\right)^{2}+\frac{25}{3}y^{2}+\frac{50}{3}y+\frac{25}{3} = 6\left(x+\frac{4}{3}-\frac{2}{3}\right)^{2}+\frac{25}{3}(y^{2}+2y+1) = 6\left(x+\frac{4}{3}-\frac{2}{3}\right)^{2}+\frac{25}{3}(y+1)^{2} = 6\left(x+\frac{1}{3}(y+1)-\frac{1}{3}-\frac{2}{3}\right)^{2}+\frac{25}{3}(y+1)^{2} = 6\left(x+\frac{1}{3}(y+1)-1\right)^{2}+\frac{25}{3}(y+1)^{2} = 6\left((x-1)+\frac{1}{3}(y+1)\right)^{2}+\frac{25}{3}(y+1)^{2} = 6(x-1)^{2}+\frac{12}{3}(x-1)(y+1)+\frac{6}{9}(y+1)^{2}+\frac{25}{3}(y+1)^{2} = 6(x-1)^{2}+4(x-1)(y+1)+\frac{27}{3}(y+1)^{2} = 6(x-1)^{2}+4(x-1)(y+1)+9(y+1)^{2}$

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 6 & 2 \\ 2 & 9 \end{pmatrix} \implies \det{\Sigma^{-1}} = 54-4 = 50$

$\Sigma = \frac{1}{\det{\Sigma^{-1}}} \cdot \begin{pmatrix} 9 & -2 \\ -2 & 6 \end{pmatrix} = \frac{1}{50} \cdot \begin{pmatrix} 9 & -2 \\ -2 & 6 \end{pmatrix} = \begin{pmatrix} \frac{9}{50} & -\frac{1}{25} \\ -\frac{1}{25} & \frac{3}{25} \end{pmatrix}$

Из $\Sigma$ матрицы получаем следующие характеристики:

$\mathbb{D}\xi = \frac{9}{50}$

$\mathbb{D}\eta = \frac{6}{50}$

$\mathop{\mathrm{cov}}\nolimits(\xi, \eta) = -\frac{1}{25}$

Найдем СКО, коэффициент корреляции и С:

$\sigma_{\xi} = \sqrt{\frac{9}{50}}$

$\sigma_{\eta} = \sqrt{\frac{6}{50}}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta)}{\sigma_{\xi} \cdot \sigma_{\eta}} = \frac{-\frac{1}{25}}{\sqrt{\frac{9}{50} \cdot \frac{6}{50}}} = \frac{-\frac{1}{25}}{\sqrt{\frac{9 \cdot 3}{25 \cdot 50}}} = \frac{-\frac{1}{25}}{\frac{3\sqrt{3}}{25\sqrt{2}}} = -\frac{\sqrt{2}}{3\sqrt{3}} = \frac{\sqrt{6}}{9}$

$C = \frac{1}{2\pi \cdot \sqrt{\frac{1}{\det(\Sigma)^{-1}}}} = \frac{1}{2\pi \cdot \sqrt{\frac{1}{50}}} = \frac{5\sqrt{2}}{2\pi}$

Тогда плотность распределения будет равна:

$P_{\xi,\eta}(x,y) = \frac{5\sqrt{2}}{2\pi} \cdot exp\left\{ -\frac{1}{2} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 6 & 2 \\ 2 & 9 \end{pmatrix}  \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix} \right] \right\}$

```python
## График плотности распределения к №1.
import numpy as np
import matplotlib.pyplot as plt

def density(x, y):
    mean = np.array([1, -1])
    covariance = np.array([[6, 2], [2, 9]])
    diff = np.stack((x, y), axis=-1) - mean
    exponent = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return 5*np.sqrt(2) / (2 * np.pi) * np.exp(exponent)

x = np.linspace(-2, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)

Z = density(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### 3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

Посчитаем определитель и найдем $\lambda$:

$\begin{vmatrix} 6-\lambda & 2 \\ 2 & 9-\lambda \end{vmatrix} = (6-\lambda) \cdot (9-\lambda)-4 = 54-6\lambda-9\lambda+\lambda^{2}-4 = \lambda^{2}-15\lambda+50 = (\lambda-5)(\lambda-10)$

Рассмотрим случай, когда $\lambda_{1}=5$:

$\left\| \overrightarrow{\mathrm{V}_{1}} \right\| = \sqrt{(-2)^{2}+1^{2}} = \sqrt{5} \implies \overrightarrow{\mathrm{V}_{1}^{'}} = \begin{pmatrix} -\frac{2}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix}$

Рассмотрим случай, когда $\lambda_{2}=10$:

$\left(\begin{array}{rr|r} -4 & 2 & 0 \\ 2 & -1 & 0 \end{array}\right) \sim \left(\begin{array}{rr|r } -2 & 1 & 0 \\ 2 & -1 & 0 \end{array}\right) \sim \left(\begin{array}{rr|r} 2 & -1 & 0 \\ 0 & 0 & 0 \end{array}\right)$

$Q = \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix}$

$Q^{\intercal} = \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix}$

$\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E}\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix} \cdot \begin{pmatrix} 1 \\ -1 \end{pmatrix} = \begin{pmatrix} -\frac{3}{\sqrt{5}} \\ -\frac{1}{\sqrt{5}} \end{pmatrix}$

$\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = Q*\Sigma_{\begin{pmatrix} \xi \\ \eta \end{pmatrix}} \cdot Q^{\intercal} = \frac{1}{50} \cdot \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix} \cdot \begin{pmatrix} 9 & -2 \\ -2 & 6 \end{pmatrix} \cdot \begin{pmatrix} -\frac{2}{\sqrt{5}} & \frac{1}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}} \end{pmatrix} = \begin{pmatrix} \frac{1}{5} & 0 \\ 0 & \frac{1}{10} \end{pmatrix} = \begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix} \implies \Sigma^{-1}_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}} = \begin{pmatrix} \lambda_{1} & 0 \\ 0 & \lambda_{2} \end{pmatrix} = \begin{pmatrix} 5 & 0 \\ 0 & 10 \end{pmatrix}$


Выпишем числовые характеристики, полученные в $\Sigma_{\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix}}$:

$\mathbb{D}_{\xi} = \frac{1}{5}; \mathbb{D}_{\eta} = \frac{1}{10}; \mathop{\mathrm{cov}}\nolimits(\xi,\eta) = 0 \implies \rho_{\xi,\eta} = 0$

$\sigma_{\xi} = \frac{1}{\sqrt{1}};   \sigma_{\eta} = \frac{1}{\sqrt{10}}$

$C = \frac{1}{2\pi \cdot \frac{1}{\sqrt{50}}} = \frac{\sqrt{50}}{2\pi}$

Запишем плотность полученного распределения: 

$P_{\xi,\eta}(x,y)= \frac{\sqrt{50}}{2\pi} \cdot exp\left\{-\frac{1}{2} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{3}{\sqrt{5}} \\ -\frac{1}{\sqrt{5}} \end{pmatrix}\right]^{\intercal} \cdot \begin{pmatrix} 5 & 0 \\ 0 & 10 \end{pmatrix} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\frac{3}{\sqrt{5}} \\ -\frac{1}{\sqrt{5}} \end{pmatrix} \right] \right\}$


```python
## График плотности распределения к №3.
import numpy as np
import matplotlib.pyplot as plt

def density(x, y):
    mean = np.array([-3/np.sqrt(5), -1/np.sqrt(5)])
    covariance = np.array([[5, 0], [0, 10]])
    diff = np.stack((x, y), axis=-1) - mean
    exponent = -0.5 * np.einsum('...k,kl,...l->...', diff, covariance, diff)
    return np.sqrt(50) / (2 * np.pi) * np.exp(exponent)

x = np.linspace(-5, 2, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = density(X, Y)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(ξ - 3η, -ξ - 5η)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = A_{4} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix}$

Найдем $\Sigma$ и числовые характеристики распределения:

$\Sigma_{4} = \frac{1}{50} \cdot \begin{pmatrix} 1 & -3 \\ -1 & 5 \end{pmatrix} \cdot \begin{pmatrix} 9 & -2 \\ -2 & 6 \end{pmatrix} \cdot \begin{pmatrix} 1 & -1 \\ -3 & -5 \end{pmatrix} = \frac{1}{50} \cdot \begin{pmatrix} 15 & -20 \\ 1 & -18 \end{pmatrix} \cdot \begin{pmatrix} 1 & -1 \\ -3 & -5 \end{pmatrix} = \frac{1}{50} \cdot \begin{pmatrix} 75 & 85 \\ 85 & 139 \end{pmatrix} = \begin{pmatrix} \frac{3}{2} & \frac{17}{10} \\ \frac{17}{10} & \frac{139}{50} \end{pmatrix}$

$\mathbb{D}_{\xi} = \frac{3}{2}$

$\sigma_{\xi} = \sqrt{\frac{3}{2}}$

$\sigma_{\eta} = \sqrt{\frac{139}{50}}$

$\rho_{\xi,\eta} = \frac{\mathop{\mathrm{cov}}\nolimits(\xi,\eta)}{\sigma_{\xi} \cdot \sigma_{\eta}} = \frac{\frac{17}{10}}{\sqrt{\frac{3 \cdot 139}{2 \cdot 50}}} = \frac{17}{\sqrt{417}}$

$\Sigma_{4}^{-1} = \frac{25}{32} \cdot \frac{1}{50} \cdot \begin{pmatrix} 75 & 85 \\ 85 & 139 \end{pmatrix} = \begin{pmatrix} \frac{75}{64} & \frac{85}{64} \\ \frac{85}{64} & \frac{139}{64} \end{pmatrix}$

$C = \frac{1}{2\pi \cdot \sqrt{\det{\Sigma_{4}}}} = \frac{1}{2\pi \cdot \sqrt{\frac{32}{25}}} = \frac{5}{8\pi \sqrt{2}}$

Запишем плотность распределения: 

$P_{\xi,\eta}(x,y) = \frac{5}{8\pi \sqrt{2}} \cdot exp \left\{-\frac{1}{2} \cdot \left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 4 \\ 4 \end{pmatrix} \right]^{\intercal} \cdot \begin{pmatrix} \frac{75}{64} & \frac{85}{64} \\ \frac{85}{64} & \frac{139}{64} \end{pmatrix} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 4 \\ 4 \end{pmatrix} \right]\right\}$

### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$.

$P_{(\xi|\eta=y)} = \frac{P_{\xi,\eta}(x,y)}{P_{\eta}(y)}$

Найдем $P_{\eta}(y)$ по формуле:

$P_{\eta}(y) = \frac{1}{\sqrt{2\pi} \cdot \sqrt{\det{\Sigma}}} \cdot exp^{-\frac{1}{2}(y-\mathbb{E}_{y}) \cdot \Sigma^{-1} \cdot (y-\mathbb{E}_{y})}$

Возьмем некоторые данные из №1:

$\mathbb{E}_{\eta} = -1$

$\mathbb{D}_{\eta} = \frac{6}{50}$

Так как рассматриваем одномерный случай, то сделовательно: 

$\Sigma_{\eta} = \mathbb{D}_{\eta} = \frac{6}{50} \implies \Sigma^{-1}_{\eta} = \frac{50}{6}$

$P_{\eta}(y) = \frac{5}{\sqrt{6\pi}} \cdot exp^{-\frac{1}{2}(y+1) \cdot \frac{50}{6}(y+1)}$

Теперь найдем условное распределение:

$P_{(\xi|\eta=y)} = \frac{\frac{5}{2\pi} \cdot exp^{-\frac{1}{2}(6x^{2}+4xy+9y^{2}-8x+14y+11)}}{{\frac{5}{\sqrt{6\pi}} \cdot exp^{-\frac{1}{2}\left((y+1) \cdot \frac{50}{6}(y+1)\right)}}} = \sqrt{\frac{3}{\pi}} \cdot \frac{exp^{-\frac{1}{2}(6x^{2}+4xy+9y^{2}-8x+14y+11)}}{exp^{-\frac{1}{2}\left(\frac{25}{3}y^{2}+\frac{50}{3}y+\frac{25}{3}\right)}} = \sqrt{\frac{3}{\pi}} \cdot exp^{-\frac{1}{2}\left(6x^{2}+4xy+9y^{2}-8x+14y+11-\frac{25}{3}y^{2}-\frac{50}{3}y-\frac{25}{3}\right)} = \sqrt{\frac{3}{\pi}} \cdot exp^{-\frac{1}{2}\left(6x^{2}+4xy-8x+\frac{2}{3}y^{2}-\frac{8}{3}y+\frac{8}{3}\right)} = \sqrt{\frac{3}{\pi}} \cdot exp^{-\frac{1}{2}\left(\frac{1}{6}\left(x^{2}+\frac{2}{3}xy-\frac{4}{3}x+\frac{1}{9}y^{2}-\frac{4}{9}y+\frac{4}{9}\right)\right)} = \sqrt{\frac{3}{\pi}} \cdot exp^{-\frac{1}{2}\left(\frac{1}{6}\left(x^{2}+2x\left(\frac{1}{3}y-\frac{2}{3}\right)+\left(\frac{1}{3}y-\frac{2}{3}\right)^{2}\right)\right)} = \sqrt{\frac{3}{\pi}} \cdot exp^{-\frac{1}{2}\left(\frac{1}{6}\left(x+\left(\frac{1}{3}y-\frac{2}{3}\right)\right)^{2}\right)}$

По формуле:

$P_{(\xi|\eta)} = C \cdot exp^{-\frac{1}{2} \cdot \left(\frac{1}{\mathbb{D}_{(\xi|\eta)}} \cdot(x-\mathbb{E}_{(\xi|\eta)})^{2} \right)}$ 

Найдем $\mathbb{E}(\xi|\eta)$ и $\mathbb{D}(\xi|\eta)$:

$P_{(\xi|\eta)} = \sqrt{\frac{3}{\pi}} \cdot exp^{-\frac{1}{2}\left(\frac{1}{6}\left(x+\left(\frac{1}{3}y-\frac{2}{3}\right)\right)^{2}\right)} \implies \mathbb{E}(\xi|\eta) = -\left(\frac{1}{3}y-\frac{2}{3}\right)^{2}; \mathbb{D}(\xi|\eta) = \frac{1}{6}$
