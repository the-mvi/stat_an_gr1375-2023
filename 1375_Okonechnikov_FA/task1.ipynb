## Основные теоретические положения:
Случайный вектор $\overrightarrow{\xi} = \left(\xi_1,...,\xi_n\right)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют **нормальным**(гауссовым) если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1,...,\eta_m\right)^\intercal$, такие что:<br>
$\overrightarrow{\xi} = \mathrm{A} \overrightarrow{\eta}+ \mathrm{B}$<br>
**Плотность нормального вектора** имеет вид:<br>
$p_{\overrightarrow{\xi}}\left(x_1,...,x_n\right) = \frac{1}{\left(2\xi\right)^\frac{n}{2}}\mathrm{exp}\left(\frac{1}{2}\left(\overrightarrow{x} - \mu \right)^{\intercal} \sigma^{-1}\left(\overrightarrow{x} - \mu \right)\right)<br>
где \overrightarrow{x} = (x_1, ..., x_n)^\intercal$<br>
**Свойства многомерного нормального распределения:**
*   Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными.
*   Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальнымы распределениями.
*   Параметры многомерного нормального распледеления(среднее и ковариационная матрица) полностью описывают его форму.
*   Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\overrightarrow{\xi} \sim \mathcal{N}\left(\mu,\Sigma\right)$, $\mathrm{A}$ - произвольная матрица размерности $m \times n$, а $\mathrm{B}$ - произвольный вектор размерности  $m \times 1$, то
$\mathrm{A}\overrightarrow{\xi} + \mathrm{B} \sim \mathcal{N}\left(\mathrm{A}\mu + \mathrm{B},\mathrm{A}\Sigma\mathrm{A}^\intercal\right) $
**Ковариационные характеристики случайного вектора:**
* Ковариационная матрица случайного вектора - это квадратная симметрическая неотрицательно определённая матрица, на диагонали которой располагаются дисперсии компонент вектор, а внедиагональные элементы - ковариации между компонентами;
* Коэффициент корреляции - это мера линейной зависимости двух случайных величин:
$\rho(x,y) = \frac{\mathrm{cov}(x, y)}{\sqrt{\mathbb{D}x\cdot\mathbb{D}y}}$, где $\mathrm{cov}(x,y)$ - ковариация, а $\mathbb{D}$-дисперсия$

Задание 1<br>
Вычислить вектор мат.ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $\mathbb{p_{\xi,\eta}}(x,y).$<br>
$\mathbb{q}(x,y) = 3x^2-6xy+11y^2+6x+10y+11=$<br>
$3x^2-6xy+6x+11y^2+10y+11=$<br>
$3(x^2-2xy+2x)+11y^2+10y+11=$<br>
$3(x^2-2x(y-1))+11y^2+10y+11=$<br>

$3(x-(y-1))^2-3(y^2-2y+1)+11y^2+10y+11=$<br>
$3(x-(y-1))^2+8y^2+16y+8=$<br>
$3(x-(y-1))^2+8(y+1)^2=$ Формула (1)<br>
$3(x-y+1-1+1)^2+8(y+1)^2=$<br>
$3((x+2)-(y+1))^2+8(y+1)^2=$<br>
$3((x+2)^2-2(x+2)(y+1)+(y+1)^2)^2+8(y+1)^2=$<br>
$3(x+2)^2-2\cdot 3(x+2)(y+1)+11(y+1)^2$<br><br>
$\mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -2 \\ -1 \end{pmatrix}$<br><br>
$\Sigma^{-1} = \begin{pmatrix} 3 & -3 \\ -3 & 11 \end{pmatrix}$<br><br>
$\Sigma = \large\frac{1}{24}\normalsize\begin{pmatrix} 11 & 3 \\ 3 & 3 \end{pmatrix}$<br><br>
$\mathbb{D}(\xi) = \large\frac{11}{24}$<br><br>
$\mathbb{D}(\eta) = \large\frac{3}{24}$<br><br>
$\mathrm{cov}(\xi,\eta) = \large\frac{3}{8}$<br><br>
$\rho=\Large\frac{\large\frac{1}{18}}{\sqrt{\large\frac{3\cdot 11}{24 \cdot 24}}}\normalsize = \large\frac{4}{3\sqrt{33}}$<br><br>
$\mathrm{det}\Sigma = \large\frac{33-9}{24\cdot 24}\normalsize = \large\frac{24}{24^2}$<br><br>
$\mathrm{C}=\large\frac{24}{2\pi\sqrt{24}}\normalsize=\large\frac{6}{\pi\sqrt{6}}$<br><br>
$\mathbb{p_{\xi,\eta}}(x,y) = \large\frac{6}{\pi\sqrt{6}}\normalsize \mathrm{exp} \Bigl\{-\large\frac{1}{2}\normalsize \Bigl [ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -2 \\ -1 \end{pmatrix}\Bigr ]^{\intercal} \begin{pmatrix} 3 & -3 \\ -3 & 11 \end{pmatrix}\Bigl [ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -2 \\ -1 \end{pmatrix}\Bigr ]\Bigr\}$

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x,y):
    #сама плотность распределения
    return (6/(np.pi*np.sqrt(6)))*np.exp((-0.5)*(3*x**2+6*x+11-6*x*y+10*y+11*y**2))
#Присваивание значений сетки графика
x = np.linspace(-4, 0, 100)
y = np.linspace(-2, 0, 100)
X,Y = np.meshgrid(x,y)
Z = f(X, Y)
#Отрисовка
fig = plt.figure(figsize=(5, 8))
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, cmap = 'copper', alpha=0.8)
plt.show()
```

Задание 3<br>
Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами.<br>
Построить график плотности полученного распределения.<br>
$\Sigma^(-1) = \begin{pmatrix} 3 & -3 \\ -3 & 11 \end{pmatrix}$<br>
Найдём матрицу ортогональных преобразований:<br>
$\begin{vmatrix} 3-\lambda & -3 \\ -3 & 11-\lambda \end{vmatrix}=\lambda^2-14\lambda+24 = 0$<br><br>
$\lambda_{1} = 2, \lambda_{2} = 12$<br>
Для $\lambda_{1} = 2$<br>
$\begin{pmatrix} 1 & -3 \\ -3 & 9 \end{pmatrix}\approx\begin{pmatrix} 1 & -3 \end{pmatrix} = 0$<br>
$x-3y = 0$<br>$x=3y$<br>
Если $y= \alpha$, тогда $x = 3\alpha$, тогда<br> $\vec{\upsilon_{1}} = \begin{pmatrix} 3\alpha \\ \alpha \end{pmatrix} \overset{\alpha = 1}{=}\begin{pmatrix} 3 \\ 1 \end{pmatrix}$ Собственный вектор<br>
$\|\vec{\upsilon_{1}}^\prime\|=\sqrt{10}$<br>
$\vec{\upsilon_{1}}^\prime = \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} 3 \\ 1 \end{pmatrix}$<br>
Для $\lambda_{2} = 12$<br>
$\begin{pmatrix} -9 & -3 \\ -3 & -1 \end{pmatrix}\approx\begin{pmatrix} 3 & 1 \end{pmatrix} = 0$<br>
$3x+y = 0$<br>$y=-3x$<br>
Если $x= \alpha$, тогда $y = -3\alpha$, тогда<br> $\vec{\upsilon_{2}} = \begin{pmatrix} \alpha \\ -3\alpha \end{pmatrix} \overset{\alpha = 1}{=}\begin{pmatrix} 1 \\ -3 \end{pmatrix}$ Собственный вектор<br>
$\|\vec{\upsilon_{2}}^\prime\|=\sqrt{10}$<br>
$\vec{\upsilon_{2}}^\prime = \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} 1 \\ -3 \end{pmatrix}$<br>
$\mathrm{Q}^{\intercal} = \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}$<br>
$\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$<br>
$\mathbb{E}\begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}\begin{pmatrix} -2 \\ -1 \end{pmatrix} = \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} -7 \\ 1 \end{pmatrix}$<br>
$\Sigma_{3} = \large\frac{1}{10}\normalsize\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}\Sigma\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}$<br>
$\Sigma_{3} = \large\frac{1}{240}\normalsize\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix}\begin{pmatrix} 11 & 3 \\ 3 & 3 \end{pmatrix}\begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} = \large\frac{1}{240}\normalsize\begin{pmatrix} 120 & 0 \\ 0 & 20 \end{pmatrix} = \begin{pmatrix} \frac{1}{2} & 0 \\ 0 & \frac{1}{12} \end{pmatrix}$<br><br>
$\Sigma_{3}^{-1} = \begin{pmatrix} 2 & 0 \\ 0 & 12 \end{pmatrix}$<br>
$\mathrm{det}\Sigma_{3} = \large\frac{1}{24}$<br>
$\mathrm{C}=\Large\frac{1}{2\pi\sqrt{\large\frac{1}{24}}}\normalsize=\large\frac{\sqrt{6}}{\pi}$<br>
$\mathbb{p_{\xi_{3},\eta_{3}}}(x,y) = \large\frac{\sqrt{6}}{\pi}\normalsize\mathrm{exp} \Bigl\{-\large\frac{1}{2}\normalsize\Bigl [\begin{pmatrix} x \\ y \end{pmatrix} - \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} -7 \\ 1 \end{pmatrix} \Bigr ]^{\intercal} \begin{pmatrix} 2 & 0 \\ 0 & 12 \end{pmatrix} \Bigl [\begin{pmatrix} x \\ y \end{pmatrix} - \large\frac{1}{\sqrt{10}}\normalsize\begin{pmatrix} -7 \\ 1 \end{pmatrix} \Bigr ] \Bigr\}$

```python
import matplotlib.pyplot as plt
import numpy as np
def f(x,y):
    #сама плотность распределения
    return (np.sqrt(6)/np.pi)*np.exp((-0.5)*(2*x**2+((14*np.sqrt(10))/5)*x+11+12*y**2-((12*np.sqrt(10))/5)*y))
#Присваивание значений сетки графика
x = np.linspace(-4, 0, 100)
y = np.linspace(1, -1, 100)
X,Y = np.meshgrid(x,y)
Z = f(X, Y)
#Отрисовка
fig = plt.figure(figsize=(5,8))
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, cmap = 'copper', alpha=0.8)
plt.show()
```
