## Основные теоретические положения
1. Случайный вектор $\vec\xi=\left(\xi_1,\dots,\xi_n\right)^{\intercal}:\Omega\rightarrow\mathbb{R}^n$ называют нормальным если существуют вектор независимых стандартных нормальных случайных величин $\vec\eta=\left(\eta_1,\dots,\eta_m\right)^{\intercal}$, матрица $A$ размерности n\times m и вещественный вектор $\vec B=\left(b_1,\dots,b_n\right)^{\intercal}$, такие что: $\vec\xi=A\vec\eta+\vec B$

2. Плотность нормального вектора имеет вид: <br>
$p_\vec\xi\left(x_1,\dots,x_n\right) = C\cdot\exp\left(-\frac12\left(\vec x-\mu\right)^{\intercal}\Sigma^{-1}\left(\vec x-\mu\right)\right)$ где:<br> 

$C=\frac1{\left(2\pi\right)^\frac n2\cdot \sqrt{det\Sigma}}$

$\mu=\mathbb{E}\vec\xi=\left(\mathbb{E}\xi_1,\dots,\mathbb{E}\xi_n\right)^{\intercal}$

$\Sigma = \left\{\sigma_{ij}\right\}_{ij=1}^n$ - ковариационная матрица

$\sigma_{ij} = \mathop{\mathrm{cov}}\nolimits(\xi_i, \xi_j)$ - ковариация

3. Свойства многомерного нормального распределения:

1)Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными

2)Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями.

3)Параметры многомерного нормального распределения(средние и ковариационная матрица) полностью описывают его форму.

4)Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec \xi \sim \mathcal{N}\left(\mu, \Sigma\right)$ - произвольная матрица размерности $m\times n$, а $B$ - произвольный вектор размерности $m\times 1$, то $A\vec\xi+\vec B \sim \mathcal{N}\left(A\vec\mu+\vec B, A\Sigma A^{\intercal}\right)$

4. Коэффицент корреляции:

$\rho\left(\xi,\eta\right)=\frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta)}{\mathbb{D}\left(\xi,\eta\right)}$

5. Формула двумерного нормального распределения величин $\xi$ и $\eta$ с математическими ожиданиями $\mathbb{E}\xi$ и $\mathbb{E}\eta$, дисперсиями $\mathbb{D}\xi$ и $\mathbb{D}\eta$ и коэффицентом корреляции $\rho$:

$p_{\xi,\eta}\left(x,y\right) = C\cdot\exp\left(\frac{-1}{2\left(1-\rho^2\right)}\left(\frac{\left(x-\mathbb{E}\xi\right)^2}{\mathbb{D}\xi}-\frac{2\rho\left(x-\mathbb{E}\xi\right)\left(y-\mathbb{E}\eta\right)}{\sqrt{\mathbb{D}\xi\cdot\mathbb{D}\eta}}+\frac{\left(y-\mathbb{E}\eta\right)^2}{\mathbb{D}\eta}\right)\right) $

6. Формула для нахождения условного распределения $\xi$ при условии $\eta$:

$p_{\xi|\eta=y} = \frac{p_{\xi,\eta}\left(x,y\right)}{p_{\eta}\left(y\right)} $
