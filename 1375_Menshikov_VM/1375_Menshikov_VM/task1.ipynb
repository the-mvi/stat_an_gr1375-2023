## Основные теоретические положения


Случайный вектор $\overrightarrow{\xi} = \left(\xi_1, \dots, \xi_n\right)^\intercal:\Omega\to \mathbb{R}^n$ имеет многомерное нормальное распределение, если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1, \dots, \eta_m\right)^\intercal$, матрица $\mathbf{A}$ размерности $n \times m$ и вещественный вектор $\overrightarrow{b} = (b_1,\dots,b_n)^\intercal$, такие что:

$\overrightarrow{\xi} = \mathbf{A}\overrightarrow{\eta} + \overrightarrow{b}$.

Плотность многомерного нормального распределения : $\mathcal{p}_{\overrightarrow{\xi}}\left(x_1, \dots, x_n\right) = \frac{1}{(2\pi)^{n/2}\sqrt{\det \Sigma}} \exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)^\intercal \Sigma^{-1}\left(\overrightarrow{x}-\overrightarrow{\mu}\right)\right)$, где $\overrightarrow{\mu} = \mathbb{E}\overrightarrow{\xi} = \left(\mathbb{E}\xi_1, \dots, \mathbb{E}\xi_n\right)^\intercal$ - вектор математических ожиданий компонент, $\Sigma = \left\{\sigma_{i,j} = \mathop{\mathrm{cov}}\left(\xi_i,\xi_j\right)\right\}_{i,j=1}^n$ - ковариационная матрица.

Свойства многомерного нормального распределения:
- Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. 
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $n \times 1$, то 
$A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

**Ковариационные характеристики случайного вектора:**,
        * Ковариационная матрица случайного вектора - это квадратная симметрическая неотрицательно определённая матрица, на диагонали которой располагаются дисперсии компонент вектор, а внедиагональные элементы - ковариации между компонентами;,
        * Коэффициент корреляции - это мера линейной зависимости двух случайных величин:,
        $p(x,y)\;=\;\frac{cov(x,y)}{\sqrt{\mathbb{D}_x\cdot \mathbb{D}_y}}$
        где ${cov(x,y)}$ - ковариация, а $\mathbb{D}$-дисперсия,

## Выполнение работы

### 1. Вычисление характеристик данного случайного вектора. Построение графика плотности распределения $\mathcal{p}_{\xi,\eta}$.

$p_{\xi,\eta}\;(x,y)\;=\;\mathrm C\;\cdot\;\exp(-\frac12(6\mathrm x^2\;+\;4\mathrm{xy}\;+\;9\mathrm y^2\;-\;8\mathrm x\;+14\mathrm y\;+11))\;$<br> 

Выполним преобразования выражения:
$6\mathrm x^2\;+\;4\mathrm{xy}\;+\;9\mathrm y^2\;-\;8\mathrm x\;+14\mathrm y\;+11\; =\\= 6(\mathrm x^2\;+\;\frac23\mathrm x(\mathrm y-2)+\frac{{(\mathrm y-2)}^2}9)\;-\frac{{2(\mathrm y-2)}^2}3\;+\;9\mathrm y^2\;-\;8\mathrm x\;+14\mathrm y\;+11 =\\= 6(\mathrm x\;+\;\frac13{(\mathrm y-2)})^2+8\frac13{(\mathrm y+1)}^2\;  =\\= 6(\mathrm x\;-\;1\;+\frac13{(\mathrm y+1)}^2)\;+8\frac13{(\mathrm y+1)}^2\;+\;1=\\=6{(\mathrm x-1)}^2\;+4(\mathrm x-1)(\mathrm y+1)\;+\;\frac{2{(\mathrm y+1)}^2}3+8\frac13{(\mathrm y+1)}^2\; =\\= 6{(\mathrm x-1)}^2\;+4(\mathrm x-1)(\mathrm y+1)\;+\;9{(\mathrm y+1)}^2\;\;$


Выпишем вектор математических ожиданий, обратную ковариационную матрицу и матрицу ковариации.

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ -1 \end{pmatrix}$
<br>
$\Sigma^{-1} = \begin{pmatrix} 6 & 2 \\ 2 & 9\end{pmatrix}$
<br>
$\Sigma = \frac{1}{50} \begin{pmatrix} 9 & -2 \\ -2 & 6\end{pmatrix}$
<br>
$\mathbb{D}\xi = \frac{9}{50},\ \mathbb{D}\eta = \frac{3}{25},\ \mathop{\mathrm{cov}}(\xi, \eta) = -\frac{1}{25}$

Найдем коэффициент корреляции и неизвестную $C$.

$\rho(\xi, \eta) = -\frac1{25}\;\cdot\;\frac5{\sqrt3}\;\cdot\;\frac{2\sqrt5}3\;=\;-\frac2{15\sqrt3}$
<br>
$C = \frac{1}{2\pi\cdot\frac{1}{\sqrt{50}}} = \frac{5\sqrt{2}}{2\pi}$

Выпишем функцию плотности распределения.

$\mathcal{p}_{\xi,\eta}(x,y) = \frac{5\sqrt{2}}{2\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix} \right]^\intercal \begin{pmatrix} 6 & 2 \\ 2 & 9\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ -1 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
mu = np.array([1, -1]) # вектор мат ожидания
cov = np.array([[9/50, -1/25], [-1/25, 3/25]]) #  матрица ковариации
inv_cov = np.linalg.inv(cov) # обратная ковариационная матрица
det_cov = np.linalg.det(cov) # det  матрицы ковариации

def pdf(x, y):
    xy = np.array([x, y]) - mu
    exponent = -0.5 * np.dot(np.dot(xy, inv_cov), xy)
    return (1 / (2 * np.pi * np.sqrt(det_cov))) * np.exp(exponent)
    
Z = np.vectorize(pdf)(X, Y)
#график
grf = plt.figure()
A = grf.add_subplot(111, projection='3d')
A.plot_surface(X, Y, Z, cmap='viridis')
A.set_xlabel('x')
A.set_ylabel('y')
plt.show()
```

Отсюда получим аффинное преобразование, переводящее исходный случайный вектор $\begin{pmatrix} \xi \\ \eta \end{pmatrix}$ в стандартный нормальный $\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix}$.

$\begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \sqrt{6} & \frac{\sqrt{6}}{3} \\ 0 & \frac{5}{\sqrt{3}}\end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} -\frac{2\sqrt{6}}{3} \\ \frac{5}{\sqrt{3}} \end{pmatrix}$

### 3. Нахождение ортогонального преобразования, переводящего соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построение графика плотности полученного распределения.

$\mathbb{D}\xi_{ind} = \frac{1}{5},\\ \mathbb{D}\eta_{ind} = \frac{1}{10},\\ \mathop{\mathrm{cov}}(\xi_{ind}, \eta_{ind}) = 0,\\ \rho(\xi_{ind}, \eta_{ind}) = 0$

$\mathcal{p}_{\xi_{ind},\eta_{ind}}(x,y) = \frac{5\sqrt{2}}{2\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{-3}{\sqrt{5}} \\ \frac{-1}{\sqrt{5}} \end{pmatrix} \right]^\intercal \begin{pmatrix} 5 & 0 \\ 0 & 10\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix}\frac{-3}{\sqrt{5}} \\ \frac{-1}{\sqrt{5}}\end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 4, 100)
y = np.linspace(-3, 4, 100)
X, Y = np.meshgrid(x, y)
mu = np.array([-3/pow(5,0.5), -1/pow(5,0.5)]) # вектор математических ожиданий 
cov = np.array([[1/5, 0], [0, 1/10]]) # ковариационная матрица
inv_cov = np.linalg.inv(cov) # обратная ковариационная матрица
det_cov = np.linalg.det(cov) # определитель ковариационной матрицы

def pdf(x, y):
    xy = np.array([x, y]) - mu
    exponent = -0.5 * np.dot(np.dot(xy, inv_cov), xy)
    return (1 / (2 * np.pi * np.sqrt(det_cov))) * np.exp(exponent)
    
Z = np.vectorize(pdf)(X, Y)
#график
grf = plt.figure()
A = grf.add_subplot(111, projection='3d')
A.plot_surface(X, Y, Z, cmap='viridis')
A.set_xlabel('x')
A.set_ylabel('y')
plt.show()
```

### 4. Вычисление характеристик совместного распределения случайного вектора $(-4\xi + 2\eta, 2\xi + \eta)^\intercal$, нахождение его плотности. Построение графика плотности полученного распределения.

Выпишем матрицу линейного преобразования.

$A = \begin{pmatrix} -3 & -2 \\ -3 & -1\end{pmatrix}$

Данная матрица — невырожденная ($\det A = -3$). 

Приступим к вычислениям

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -3 & -2 \\ -3 & -1\end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$
<br>
$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -3 & -2 \\ -3 & -1\end{pmatrix} \begin{pmatrix} 1 \\ -1 \end{pmatrix} = \begin{pmatrix} -1 \\ -2 \end{pmatrix}$
<br>
$\Sigma_{4} = \frac{1}{50}\begin{pmatrix} -3 & -2 \\ -3 & -1\end{pmatrix}\begin{pmatrix} 9 & -2 \\ -2 & 6\end{pmatrix}\begin{pmatrix} -3 & -3 \\ -2 & -1\end{pmatrix} = \frac{1}{50}\begin{pmatrix} 81 & 75 \\ 75 & 75\end{pmatrix}$
<br>
$\mathbb{D}\xi_{4} = \frac{81}{50},\\ \mathbb{D}\eta_{4} = \frac{3}{2},\\ \mathop{\mathrm{cov}}(\xi_{4}, \eta_{4}) = \frac{3}{2},\\ \rho(\xi_{4}, \eta_{4}) =  -\frac{5\sqrt{3}}{9}$
<br>
$\det \Sigma_4 = \frac{1}{50\cdot50}(81\cdot75 - 75^2) = \frac{9}{50}\\
\Sigma_{4}^{-1} = \frac{50}{9} \cdot \frac{1}{50} \begin{pmatrix} 75 & -75 \\ -75 & 81\end{pmatrix} = \begin{pmatrix} \frac{75}{9} & \frac{-75}{9} \\ \frac{-75}{9} & 9\end{pmatrix}$
<br> 
$C = \frac{1}{2\pi\sqrt{\frac{9}{50}}} = \frac{5\sqrt{2}}{6\pi}$

Выпишем плотность полученного распределения.

$\mathcal{p}_{\xi_4,\eta_4}(x,y) = \frac{5\sqrt{2}}{6\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ -2 \end{pmatrix} \right]^\intercal \begin{pmatrix} \frac{75}{9} & \frac{-75}{9} \\ \frac{-75}{9} & 9\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ -2 \end{pmatrix} \right]\right\}$
