## Основные теоретические положения
1. Случайный вектор $ \overrightarrow{\xi} = (\xi_1,...,\xi_n)^\intercal: \Omega\rightarrow\mathbb{R}^n $ называют *нормальным(гауссовым)* если существует вектор независимых нормальных случайных величин $\overrightarrow{\eta} = (\eta_1,...,\eta_m)^\intercal$ , матрица $A$ размерности $n \times m$ и вещественный вектор $B = (b_1,...,b_m)^\intercal$, такие что: $$\overrightarrow{\xi} = A\overrightarrow{\eta} + B$$
  * Обозначение: $\overrightarrow{\xi} \sim \mathcal{N}(\mu, \Sigma)$, где $\mathbb{E}\overrightarrow{\xi} = (\mathbb{E}\xi_1,...,\mathbb{E}\xi_n)^\intercal$ - вектор математических ожиданий компонент; $\Sigma = \{\sigma_{ij}\}_{i,j=1}^n$, $\sigma_{ij} = \mathrm{cov}(\xi_i, \xi_j)$ - ковариационная матрица.
  * Плотность нормального вектора имеет вид: $$ p_\overrightarrow{\xi}(x_1,...x_n) = \frac{1}{(2\pi)^{n/2}\sqrt{\det\Sigma}}\exp\left(-\frac{1}{2}(\overrightarrow{x}-\mu)^\intercal\Sigma^{-1}(\overrightarrow{x}-\mu)\right)$$, где $\overrightarrow{x} = (x_1,...x_n)^\intercal$
  
2. Свойства многомерного нормального распределения:
  * Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределёнными.
  * Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределёнными.
  * Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
  * Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\overrightarrow{\xi}\sim \mathcal{N}(\mu,\sigma),A $ - произвольная матрица размерности $m\times n$, а $B$ - произвольный вектор размерности $m\times 1$, то $A\overrightarrow{\xi} + B \sim \mathcal{N}(A\mu,A\Sigma A^\intercal)$

Плотность двумерного нормального распределения имеет вид:

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi,\eta}$.
1. Преобразование Лагранжа

   $q(x,y) = 5x^2-4xy+8y^2+6x+12y+9 = 5\left(x - \frac{3}{5}y + \frac{3}{5}\right)^2 + \frac{36}{5}(y + 1)^2 = 5(x+1)^2-2 \cdot 2(x+1)(y+1)+8(y+1)^2$
  
2. Математическое ожидание

   $\mathbb{E}\left( {\begin{array}{c} \xi \\ \eta  \\ \end{array} } \right) = \left( {\begin{array}{c} -1 \\ -1  \\ \end{array} } \right)$

3. Ковариационная матрица

   $ \Sigma^{-1} =  \left( {\begin{array}{cc} 5 & -2 \\ -2 & 8  \\ \end{array} } \right) \; \Sigma =  \frac{1}{36}\left( {\begin{array}{cc} 8 & 2 \\ 2 & 5  \\ \end{array} } \right)$

4. Дисперсия
  
   $\mathbb{D}(\xi) = \frac{2}{9} \;\; \mathbb{D}(\eta) = \frac{5}{36}$

5. Ковариация 

   $\mathop{\mathrm{cov}}\nolimits(\xi, \eta) = \frac{1}{18}$
  
6. Коэффициент корреляции

   $\rho(\xi, \eta) = \frac{\mathop{\mathrm{cov}}\nolimits(\xi, \eta)}{\sqrt{\mathbb{D}(\xi)\mathbb{D}(\eta)}} = \frac{\sqrt{10}}{10}$
  
7. Плотность распределения
  
   $C = \frac{1}{(2\pi)^{n / 2} \sqrt{\det\Sigma}} = \frac{6}{2\pi}$
  
   $p_{\xi,\eta}(x,y) = \frac{6}{2\pi} \exp\left(-\frac{1}{2}\left[\left( {\begin{array}{c} x \\ y  \\ \end{array} } \right)-\left( {\begin{array}{c} -1 \\ -1  \\ \end{array}} \right)\right]^\intercal\left( {\begin{array}{cc} 5 & -2 \\ -2 & 8  \\ \end{array} } \right)\left[\left( {\begin{array}{c} x \\ y  \\ \end{array} } \right)-\left( {\begin{array}{c} -1 \\ -1  \\ \end{array} } \right)\right] \right)$

```python
import matplotlib.pyplot as plt
import numpy as np
 
#функция плотности распределения
def densityFunc(x, y):
    return (6/(2*np.pi))*np.exp(-(5*x**2-4*x*y+8*y**2+6*x+12*y+9)/2)


mean = np.array([-1, -1]) #вектор мат ожиданий
eps = 2 #интервал для построения

#заполнение массивов значений
x = np.linspace(mean[0] - eps, mean[0] + eps, num = 100)
y = np.linspace(mean[1] - eps, mean[1] + eps, num = 100)
X, Y = np.meshgrid(x, y)
p = densityFunc(X, Y)

#отрисовка графика
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('График плотности распределения', fontsize=16)
ax.plot_surface(X, Y, p, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p')

plt.show()
```

### 3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$\Sigma^{-1} = \left( {\begin{array}{cc} 5 & -2 \\ -2 & 8\end{array} } \right)$

1. Собственные числа

    $\det(\Sigma^{-1} - E\lambda) = \lambda^2 - 13\lambda + 36 = 0 $

    $\lambda_1 = 4 \; \lambda_2 = 9$

2. Собственные векторы
  
  * $ \left( {\begin{array}{cc} 1 & -2 \\ -2 & 4\end{array} } \right) \sim (1 \; -2) = 0 $
 
    $\overrightarrow{v_1} = \left( {\begin{array}{c} 2 \\ 1\end{array} } \right) \;\; \overrightarrow{v'_1} = \frac{1}{\sqrt{5}}\left( {\begin{array}{c} 2 \\ 1\end{array} } \right)$

 * $ \left( {\begin{array}{cc} -4 & -2 \\ -2 & -1\end{array} } \right) \sim (-2 \; -1) = 0 $
 
    $\overrightarrow{v_2} = \left( {\begin{array}{c} 1 \\ -2\end{array} } \right) \;\; \overrightarrow{v'_2} = \frac{1}{\sqrt{5}}\left( {\begin{array}{c} 1 \\ -2\end{array} } \right)$
      
3. Матрица ортогональных преобразований

   $ Q^\intercal = \frac{1}{\sqrt{5}} \left( {\begin{array}{cc} 2 & 1 \\ 1 & -2\end{array} } \right)$

   $ \left( {\begin{array}{c} \xi_3 \\ \eta_3\end{array} } \right) = \frac{1}{\sqrt{5}} \left( {\begin{array}{cc} 2 & 1 \\ 1 & -2\end{array} } \right) \left( {\begin{array}{c} \xi \\ \eta\end{array} } \right) $

4. Математическое ожидание

   $ \mathbb{E}\left( {\begin{array}{c} \xi_3 \\ \eta_3\end{array} } \right) = \frac{1}{\sqrt{5}} \left( {\begin{array}{cc} 2 & 1 \\ 1 & -2\end{array} } \right) \left( {\begin{array}{c} -1 \\ -1\end{array} } \right) = \frac{1}{\sqrt{5}}\left( {\begin{array}{c} -3 \\ 1 \end{array} } \right) $

5. Ковариационная матрица

   $ \Sigma_3 = \frac{1}{36} \cdot \frac{1}{5} \left( {\begin{array}{cc} 2 & 1 \\ 1 & -2\end{array} } \right) \left( {\begin{array}{cc} 8 & 2 \\ 2 & 5\end{array} } \right) \left( {\begin{array}{cc} 2 & 1 \\ 1 & -2\end{array} } \right) = \left( {\begin{array}{cc} \frac{1}{\lambda_1} & 0 \\ 0 & \frac{1}{\lambda_2}\end{array} } \right) = \left( {\begin{array}{cc} \frac{1}{4} & 0 \\ 0 & \frac{1}{9}\end{array} } \right) $

   $ \Sigma_3^{-1} = \left( {\begin{array}{cc} 4 & 0 \\ 0 & 9\end{array} } \right) $

6. Плотность распределения

   $ C = \frac{6}{2\pi} $
   
   $p_{\xi,\eta}(x,y) = \frac{6}{2\pi} \exp\left(-\frac{1}{2}\left[\left( {\begin{array}{c} x \\ y  \\ \end{array} } \right)-\frac{1}{\sqrt{5}}\left( {\begin{array}{c} -3 \\ 1 \end{array} } \right)\right]^\intercal\left( {\begin{array}{cc} 4 & 0 \\ 0 & 9 \\ \end{array} } \right)\left[\left( {\begin{array}{c} x \\ y  \\ \end{array} } \right)-\frac{1}{\sqrt{5}}\left({\begin{array}{c} -3 \\ 1 \end{array} } \right)\right]\right)$

```python
import matplotlib.pyplot as plt
import numpy as np
 
#функция плотности распределения
def densityFunc(x, y):
    return (6/(2*np.pi))*np.exp(-(4*x**2+(24/5**(1/2))*x+9+9*y**2-(18/5**(1/2))*y)/2)


mean = np.array([-3/5**(1/2), 1/5**(1/2)]) #вектор мат ожиданий
eps = 2 #интервал для построения

#заполнение массивов значений
x = np.linspace(mean[0] - eps, mean[0] + eps, num = 100)
y = np.linspace(mean[1] - eps, mean[1] + eps, num = 100)
X, Y = np.meshgrid(x, y)
p = densityFunc(X, Y)

#отрисовка графика
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('График плотности распределения', fontsize=16)
ax.plot_surface(X, Y, p, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p')

plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(-2\xi-2\eta, 3\xi-\eta)^\intercal$ и записать его плотность. Построить график плотности полученного распределения.

$A = \left( {\begin{array}{cc} -2 & -2 \\ 3 & -1\end{array} } \right) \;\; \det A = 8 \neq 0$

$ \left( {\begin{array}{c} \xi_4 \\ \eta_4 \end{array} } \right) = \left( {\begin{array}{cc} -2 & -2 \\ 3 & -1\end{array} } \right) \left( {\begin{array}{c} \xi \\ \eta\end{array} } \right) $

1. Математическое ожидание 

    $ \mathbb{E}\left( {\begin{array}{c} \xi_4 \\ \eta_4 \end{array} } \right) = \left( {\begin{array}{cc} -2 & -2 \\ 3 & -1\end{array} } \right) \left( {\begin{array}{c} -1 \\ -1 \end{array} } \right) = \left( {\begin{array}{c} 4 \\ -2 \end{array} } \right)$

2. Ковариационная матрица
    $ \Sigma_4 = \frac{1}{36} \left( {\begin{array}{cc} -2 & -2 \\ 3 & -1\end{array} } \right) \left( {\begin{array}{cc} 8 & 2 \\ 2 & 5\end{array} } \right) \left( {\begin{array}{cc} -2 & 3 \\ -2 & -1\end{array} } \right) = \left( {\begin{array}{cc} \frac{17}{9} & -\frac{23}{18} \\ -\frac{23}{18} & \frac{65}{36}\end{array} } \right) $

    $ \Sigma_4^{-1} = \frac{9}{16} \left( {\begin{array}{cc} \frac{65}{36} & \frac{23}{18} \\ \frac{23}{18} & \frac{17}{9}\end{array} } \right) $ 

3. Дисперсия

    $\mathbb{D}(\xi_4) = \frac{17}{9} \;\; \mathbb{D}(\eta_4) = \frac{65}{36}$
    
4. Ковариация 

    $\mathop{\mathrm{cov}}\nolimits(\xi_4, \eta_4) = -\frac{23}{18}$

5. Плотность распределения 

    $C = \frac{3}{8\pi}$
    
    $p_{\xi,\eta}(x,y) = \frac{3}{8\pi} \exp\left(-\frac{1}{2} \cdot \frac{9}{16}\left[\left( {\begin{array}{c} x \\ y  \\ \end{array} } \right)-\left( {\begin{array}{c} 4 \\ -2 \end{array} } \right)\right]^\intercal\left( {\begin{array}{cc} \frac{65}{36} & \frac{23}{18} \\ \frac{23}{18} & \frac{17}{9}\end{array} } \right) \left[\left( {\begin{array}{c} x \\ y  \\ \end{array} } \right)-\left( {\begin{array}{c} 4 \\ -2 \end{array} } \right)\right]\right)$

```python
import matplotlib.pyplot as plt
import numpy as np
 
#функция плотности распределения
def densityFunc(x, y):
    return (3/(8*np.pi))*np.exp(-(65/64*x**2-21/4*x+9+23/16*x*y-3/2*y+17/16*y**2)/2)


mean = np.array([4, -2]) #вектор мат ожиданий
eps = 4 #интервал для построения

#заполнение массивов значений
x = np.linspace(mean[0] - eps, mean[0] + eps, num = 100)
y = np.linspace(mean[1] - eps, mean[1] + eps, num = 100)
X, Y = np.meshgrid(x, y)
p = densityFunc(X, Y)

#отрисовка графика
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('График плотности распределения', fontsize=16)
ax.plot_surface(X, Y, p, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p')

plt.show()
```
