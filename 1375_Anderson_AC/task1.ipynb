## Основные теоретические положения

Случайный вектор $\vec{\xi} = (\xi_1, ..., \xi_n)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют      $\textit{нормальным}$ (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1, ..., \eta_m)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\vec{b} = (b_1, ..., b_m)^\intercal$, такие что:

$\vec{\xi} = A\vec{\eta} + \vec{b}$

$\bullet$ Обозначение: $\vec{\xi} \sim \mathcal{N}(\xi, \Sigma)$, где $\mu = \mathrm{E}_{\vec{\xi}} = (\mathrm{E}_{\xi_1}, ..., \mathrm{E}_{\xi_n})^\intercal$ - вектор математических ожиданий компонент; $\Sigma = \{ \sigma{{}_i}_j \}_{i,j=1}^n$, $ \sigma{{}_i}_j =  \mathrm{cov}(\xi_i, \xi_j)$ - коваривационная матрица

$\bullet$ Плоность нормального вектора имеет вид:

$\mathrm{p}_{\vec{\xi}}(x_1, ..., x_n) = \frac1{(2\pi)^{n/2} \sqrt{\det{\Sigma}}} \exp\bigg(-\frac12 (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \bigg)$,

где $\vec{x} = (x_1, ..., x_n)^\intercal$

#### Свойства 


$\odot$ Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределёнными.

$\odot$ Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями.

$\odot$ Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.

$\odot$ Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}\left(\vec{\mu}, \Sigma\right)$, $A$ - произвольная неособенная матрица размерности $m \times n$, а $B$ - произвольный вектор размерности $m \times 1$, то 

$A\vec{\xi} + B \sim \mathcal{N}\left( A\vec{\mu}+B, A \Sigma A^\intercal \right)$.

---

$ \rho(\xi, \eta) = \frac{\mathrm{cov}(\xi, \eta)}{\sqrt{\mathrm{D}(\xi) \mathrm{D}(\eta)}} - \text{Формула для нахождения коэффициента корреляции}$

$ \mathrm{C} = \frac1{(2\pi)^\frac{n}2 \sqrt{\det{\Sigma}}} - \text{Формула для нахождения константы}$

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $\mathrm{p}_{\xi,\eta}$.


$q(x, y) = 7x^2 - 8xy + 13y^2 - 6x - 18y + 12 =(7x^2 - 8xy - 6x) +13y^2 - 18y + 12 =$

$= (7x^2 - 2x(4y + 3)) + 13y^2 - 18y + 12 = 7\bigg(x^2 + \big(-\frac{2x(4y + 3)}7\big) + \frac{(4y + 3)^2}{49}\bigg) - \frac{(4y + 3)^2}7 + 13y^2 - 18y + 12 =$

$= 7\bigg(x - \frac{4}7y - \frac{3}7\bigg)^2 +\frac{75y^2 - 150y + 75}7 = 7\bigg(x - \frac{4}7y - \frac{3}7\bigg)^2+ \frac{75}7(y-1)^2 = 7\bigg((x-1) - \frac{4}7(y-1)\bigg)^2 + \frac{75}7(y-1)^2 = $

$= 7\bigg((x-1)^2 - 2(x-1)\big(-\frac{4}7(y-1)\big) + \frac{16}{49}(y-1)^2\bigg) = 7(x-1)^2 - 2 \cdot 4(x-1)(y-1) + 13(y-1)^2$

$ \underset{\underset{\scriptstyle\text{Вектор мат. ожидания }}{\scriptstyle\downarrow}}{E \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 1 \\ 1 \end{pmatrix}}  $
$ \underset{\underset{\scriptstyle\text{Обратная матрица ковариации }}{\scriptstyle\downarrow}}{\Sigma^{-1} = \begin{pmatrix} 7 \ \ \ -4 \\ -4 \ \ \ 13 \end{pmatrix}}$

$ \underset{\underset{\scriptstyle\text{Матрица ковариации }}{\scriptstyle\downarrow}}{\Sigma = \frac1{75}\begin{pmatrix} 13 \ \ \ 4 \\ 4 \ \ \ 7 \end{pmatrix}}$

##### Числовые характеристики случайного вектора:

$\mathrm{D}(\xi) = \frac{13}{75} \ \  \mathrm{D}(\eta) = \frac{7}{75} - \text{ [Дисперсия для} \ \xi \ и \ \eta]$

$\mathrm{cov}(\xi, \eta) = \frac4{75} - \text{[Ковариация]}$

 $\rho(\xi, \eta) = \frac{\frac4{75}}{\sqrt{\frac{13}{75} \cdot \frac7{75}}} = \frac4{\sqrt{91}} - \text{[Коэффициент корреляции]}$
 
 $ \mathrm{C} =  \frac1{(2\pi) \sqrt{\frac1{75}}} =\frac{\sqrt{75}}{2\pi} - \text{[n = 2 т.к. двумерное распредение]}$
 
 ##### Плотность распределения в матричной записи:
 
 $\mathrm{p}_{\xi, \eta}(x, y) = \frac{\sqrt{75}}{2\pi} \exp{\biggl[ -\frac{1}2 \biggl[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ 1 \end{pmatrix} \biggr]^\intercal \begin{pmatrix} 7 \ \ -4 \\ -4 \ \ 13 \end{pmatrix} \biggl[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 1 \\ 1 \end{pmatrix} \biggr] \biggr]}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Заданные параметры
mean = np.array([1, 1])
cov_matrix = np.array([[13, 4], [4, 7]])

# Создаем сетку точек для построения графика
x, y = np.meshgrid(np.linspace(-10, 10, 500), np.linspace(-10, 10, 500))
pos = np.dstack((x, y))

# Создаем многомерное нормальное распределение с заданными параметрами
rv = multivariate_normal(mean, cov_matrix)

# Вычисляем значения плотности вероятности для каждой точки на сетке
pdf = rv.pdf(pos)

# Создаем трехмерный график
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, pdf)

ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```
