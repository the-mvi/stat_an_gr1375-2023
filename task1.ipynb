## Основные теоретические положения
Случайный вектор $\vec{\xi}=(\xi_{1},\ldots,\xi_{n})^\intercal \colon \Omega \to \mathbb{R}^n $ называют нормальным, если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta}= (\eta_{1},\ldots,\eta_{m})^\intercal$, матрица $A$ размерности $m \times n$ и вещественный вектор $B=(b_{1},\ldots,b_{n})^\intercal$, такие что $\vec{\xi}=A\vec{\eta}+B$ <br>

Коэффициент корреляции: <br>

$ \rho(\xi,\eta)=\frac{\mathop{\mathrm{cov}}\nolimits(\xi,\eta)}{\sqrt{ \mathbb{D}(\xi) \cdot \mathbb{D}(\eta)}}$ <br>
Ковариационная матрица<br>
$\Sigma =
\begin{pmatrix}
    \mathbb{D}\xi & \mathop{\mathrm{cov}}\limits(\xi, \eta) \\
    \mathop{\mathrm{cov}}\limits(\eta, \xi) & \mathbb{D}\eta
\end{pmatrix}$

Плотность нормального вектора имеет вид:

$p_{\vec{\xi}(x_{1},\ldots,x_{n})= \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{det\Sigma}} \exp\left(-\frac{1}{2}(\vec{x}-\mu)^\intercal \cdot \Sigma^{-1} \cdot (\vec{x}-\mu)\right)$,где $\vec{x}=(x_{1},\ldots,x_{n})^\intercal$

Плотность условного распределения для $\xi$: $ p(\xi|\eta=y) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределённой.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределёнными.
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное  нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}(\mu, \sigma), A$ - произвольная матрица размерности $m \times n$,а B- произвольный вектор размерности $m \times 1$,то <br>
 $$A\vec{\xi}+B \sim \mathcal{N}(A\mu, A\Sigma A^\intercal) $$

#### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.
$ q(x,y)=3x^2-3xy+7y^2+15x+5y+25 $ <br> <br>
Используем метода Лагранжа: <br> <br>
$ q(x,y)=3\left(x^2-2x\left(\frac{y-5}{2}\right)\right)+7y^2+5y+25=3\left(x^2-2x\left(\frac{y-5}{2}\right)+\frac{(y-5)^{2}}{4}\right)+7y^2+5y+25-\frac{3}{4}(y-5)^{2}=\\=3\left(x-\frac{y-5}{2}\right)^2+\frac{25}{4}y^2+\frac{50}{4}y+\frac{25}{4}= 3\left(x-\frac{y-5}{2}\right)^2+\frac{25}{4}(y+1)^2 = 3\left(x+3-\frac{1}{2}\left(y+1\right)\right)^2+\frac{25}{4}\left(y+1\right)^2 =\\= 3(x+3)^2-3(x+3)(y+1)+7(y+1)^2$
<br>

Вектор математических ожиданий:
<br>
$\mathbb{E}\begin{pmatrix}
     \xi \\
     \eta

     \begin{pmatrix}

Обратная ковариационная матрица:
<br>
$\Sigma^{-1}= \begin{pmatrix}

<br>
Матрица ковариаций:
<br>
$\Sigma = \frac{4}{75} \cdot \begin{pmatrix}
     7 & \frac{3}{2}\\

$\mathbb{D}(\xi)=\frac{4}{75} \cdot 7 = \frac{28}{75}$ <br>
$\mathbb{D}(\eta)=\frac{4}{75} \cdot 3 = \frac{4}{25}$ <br>

$\mathop{\mathrm{cov}}\nolimits(\xi,\eta)=\frac{3 \cdot 2}{75}=\frac{2}{25}$ <br>

$ \rho(\xi,\eta)=\frac{\frac{2}{25}}{\sqrt{ \frac{28}{75} \cdot \frac{4}{25}}}=\frac{\sqrt{21}}{14}$ <br>

$p_{\xi,\eta}(x,y)=\frac{5\sqrt{3}}{4\pi} \cdot \exp \left\{-\frac{1}{2}

```python
#Плотность двумерного нормального распределения:
import numpy as np
import matplotlib.pyplot as plt

mu = np.array([-3, -1])

x = np.linspace(mu[0]-2, mu[0] + 2, 150)
y = np.linspace(mu[1]-2, mu[1] + 2, 150)
X, Y = np.meshgrid(x, y)

C = 5*np.sqrt(3)/(4*np.pi)
Z = C * np.exp(-1/2*(3*X**2+15*X+25-3*X*Y+5*Y+7*Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='Greens')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('График плотности распределения $p_{\\xi,\\eta}(x,y)$')

plt.show()
```

#### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E} ( \xi | \eta )$ и $\mathbb{D} ( \xi | \eta )$
Плотность условного распределения $\xi$ найдем по следующей формуле: $ p(\xi|\eta=y) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$
Из пункта 1:<br>
$ \mathbb{E}{\eta} = -1$<br>
$ \mathbb{D}{\eta} = \frac{4}{25} $\
$p_{\eta}(y) = \frac{1}{\sqrt{2\pi} \cdot \sqrt{\mathbb{D}{\eta}}} \cdot  \exp\left\{-\frac{1}{2} \cdot \frac{(y-\mathbb{E}_{\eta})^2}{\mathbb{D}_{\eta}}\right\} = \frac{1}{\sqrt{2\pi} \cdot \sqrt{\frac{4}{{25}}}} \cdot  \exp\left\{-\frac{1}{2} \cdot \frac{(y+1)^2}{\frac{4}{25}}\right\} $
$p(\xi|\eta=y) = \frac{5\sqrt3}{4\pi} \cdot \frac{4\pi}{5\sqrt{2\pi}} \cdot \frac{ \exp\left\{-\frac{1}{2} \cdot (3x^2-3xy+7y^2+15x+5y+25)\right\}}{\exp\left\{-\frac{1}{2} \cdot 25 \frac{((y+1)^2)}{4}\right\}} =\frac{\sqrt{6\pi}}{2\pi} \cdot \frac{ \exp\left\{-\frac{1}{2} \cdot (3\left(x-\frac{y-5}{2}\right)^2 + \frac{25}{4}(y+1)^2)\right\}}{\exp\left\{-\frac{1}{2} \cdot  \frac{25}{4}((y+1)^2)\right\}} =\\= \frac{\sqrt{6\pi}}{2\pi} \cdot \exp\left\{-\frac{1}{2} \cdot \left(3\cdot \left(x-\frac{y-5}{2}\right)^2\right)\right\}$
Получаем:<br>
$\mathbb{E}(\xi|\eta=y) = \frac{y-5}{2}$<br>
$\mathbb{D}(\xi|\eta=y) = \frac{1}{3} $
