## Основные теоретические положения
#### Многомерное нормальное распределение
*Многомерное нормальное распределение является предельным для распределений нормированных сумм случайных векторов. В частности, распределение (мультиномиальное) вектора нормированных частот*$\left( n_{1}^*,\ldots,n_{N-1}^* \right)=\vec{n_{}^*},n_{i}^*=\left(\upsilon_{i}-np_{i}\right)/\sqrt{n}$ *стремиться к вырожденному нормальному распределению*$N\left(0,\Sigma=\lVert \sigma_{i,j} \rVert_{1}^{N-1}\right)$, где $\sigma_{i,j}=p_{i}\left(1-p_{i}\right),\sigma_{i,j}=p_{i}p_{j},i \neq j$ *(ковариации мульномиального распределения). Известно также, что квадратичная форма* $Y^{T}\Sigma^{-1}Y$ *, стоящая в показателе экспоненты, показывающей плотность невырожденного нормального распределения* $N\left(0,\Sigma\right)$ *, распределена по закону* $\mathcal{X}_(N-1)^2.^1$ *.Данное свойство лежит в основе* $\mathcal{X}^2$ *критерия*
#### Семейство нормальных распределений
* Обозначение: $ \xi \sim N \left(a,\sigma^2\right)$
* Параметр распределения: $\theta= \left(a,\sigma^2\right)$
* Область определения параметра: $\Theta=\mathbb{R} \times \left(0,+\infty\right)$
* Плотность распределения: $P_{\theta} \left(x\right)=\frac{1}{\sqrt{2\pi\sigma^2}} exp\left(-\frac{\left(x-a\right)^2}{2\sigma^2}\right)$
* Носитель: supp $\xi=\mathbb{R}$
* Числовые характеристики:
  * $\mathbb{E}_{\theta}\xi=a$
  * $\mathbb{D}_{\theta}\xi=\sigma^2$
  * $Skew_{\theta}\xi=Kurt_{\theta}\xi=0$
#### Распределения, получаемые из нормального распределения
*Пусть $\xi_{1},\xi_{2},\ldots,\xi_{n}$ - выборка из стандарного нормального распределения. Тогда распределение случайной величины $\mathcal{X}_{n}^2=\sum_{i=1}^n \xi_i^2$ называется хи-квадрат $\left(\mathcal{X}^2\right)$ с n-степенями свободы*
#### Распределение $\mathcal{X}^2$
* Обозначение: $ \xi \sim \mathcal{X}_n^2$
* Плотность распределения: $P_{n} \left(x\right)=\frac{1}{\Gamma \left(n/2\right) 2^{n/2}} x^{n/2-1} exp\left(-x/2\right) \mathbb{1}_{\left(x \gneq 0\right)}$
* Носитель: supp $\xi=\mathbb{R}_{+}$
* Числовые характеристики:
  * $\mathbb{E}_{\theta} \mathcal{X}_n^2=n$
  * $\mathbb{D}_{\theta} \mathcal{X}_n^2=2n$
  * $Skew_{\theta} \mathcal{X}_n^2= \sqrt{8/n}$
  * $Kurt_{\theta} \mathcal{X}_n^2=12/n$
* Хи-квадрат распределение - частный случай гамма-распределения: $\mathcal{X}_n^2 \sim Gamma\left(n/2,2\right)$
#### Аффинное преобразование:
* Аффинное преобразование $ f:\mathbb{R}^n \rightarrow \mathbb{R}^n$ есть преобразование вида:
  * $ f \left(x\right)=M \cdot x + \upsilon,$
  * где M - обратимая матрица $\upsilon \in \mathbb{R}^n$
#### Определение нормального(гауссовского) вектора* Случайный вектор $\vec{\xi}=\left(\xi_1,\ldots,\xi_n\right)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta}=\left(\eta_1,\ldots,\eta_n \right)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\vec{b} = \left(b_1,\ldots,b_n \right)^\intercal$, такие что $\vec{\xi} = A\vec{\eta} + \vec{b}$
* Обозначение: $\vec{\xi} \sim \mathcal{N} \left(\mu, \Sigma \right)$
  * где $ \mu = \mathbb{E}\vec{\xi}=\left(\mathbb\xi_1,\ldots,\mathbb\xi_n \right)^{\intercal}$ - вектор математических ожиданий компонент;
* $\Sigma = \{ \mathrm{cov} \left(\xi_i, \xi_j \right) \}_{i,j=1,1}^n $ - ковариационная матрица.
* Плотность распределения нормального случайного вектора имеет вид:
  * $p_{\vec{\xi}}\left(x_1,\ldots, x_n \right) = \frac1{\left(2\pi \right)^{\frac{n}{2}} \sqrt{\det{\Sigma}}} \exp\left(-\frac{1}{2} (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \right)$

#### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.
$q\left(x,y\right)=2x^2+4xy+5y^2+6y+3=\left(2x^2+4xy\right)+5y^2+6y+3=2\left(x^2+2xy\right)+5y^2+6y+3=2\left(x^2+2xy+y^2\right)-2y^2+5y^2+6y+3=2\left(x+y\right)^2+3\left(y^2+2y+1\right)=2\left(x+\left(y+1\right)-1\right)^2+3\left(y+1\right)^2=2\left(\left(x-1\right)^2+2\left(x-1\right)\left(y+1\right)+\left(y+1\right)^2\right)+3\left(y+1\right)^2=2\left(x-1\right)^2+2*2\left(x-1\right)\left(y+1\right)+5\left(y+1\right)^2$
###### Вектор матрицы ожиданий:
* $ \mathbb{E}  \begin{pmatrix} \xi \\ \eta \end {pmatrix} = \begin{pmatrix} 1 \\ -1 \end{pmatrix} $ 
###### Ковариационная матрица:
* $\Sigma^{-1} = \begin{pmatrix} 2 & 2 \\ 2 & 5 \end{pmatrix} \Rightarrow \Sigma = \frac{1}{6}\begin{pmatrix} 5 & -2 \\ -2 & 2 \end{pmatrix}$ - ковариация
###### Ковариационные характеристики:
* $D\left(\xi\right)=\frac{5}{6}$
* $D\left(\eta\right)=\frac{2}{6}=\frac{1}{3}$
* $cov\left(\xi,\eta\right)=-\frac{2}{6}=-\frac{1}{3}$
* $\rho=-\frac{1}{3\sqrt{\frac{5*1}{6*3}}}=-\frac{1*\sqrt{2}}{\sqrt{15}}$ - кореляция
* $C=\frac{1}{2\pi^{\frac{n}{2}}\sqrt{det\xi}}=\frac{1*\sqrt{6}}{2\pi*1}=\frac{\sqrt{6}}{2\pi}$
###### Плотность нормального распределения:
* $P_{\xi,\eta}\left(x,y\right)=\frac{\sqrt{6}}{2\pi}exp\left\lbrace -\frac{1}{2} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} 1 \\ -1 \end{pmatrix} \right\rbrack^{\intercal} \begin{pmatrix} 2 & 2 \\ 2 & 5 \end{pmatrix} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} 1 \\ -1 \end{pmatrix} \right\rbrack \right\rbrace$

```python
import plotly.graph_objects as go
import numpy as np
x, y = np.meshgrid(np.linspace(-3, 5, 100), np.linspace(-5, 5, 100))
z = (np.sqrt(6)/(2*np.pi))*np.exp(-1/2*(2*(x-1)**2+4*(x-1)*(y+1)+5*(y+1)**2))

fig = go.Figure(data=go.Surface(x=x, y=y, z=z, colorscale="Hot"))  

fig.update_layout(
    title="График функции плотности совместного распределения случайного вектора",
    scene=dict(
        xaxis=dict(title="X"),
        yaxis=dict(title="Y"),
        zaxis=dict(title="Z")
    )
)
fig.show()
```

#### 4. Вычислить характеристики совместного распределения случайного вектора $\left(\xi - 3 \eta,3 \xi - 3 \eta \right)^{\intercal} $ и записать его плотность.
* $\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}=\begin{pmatrix} \xi - 3\eta \\ 3\xi + 3\eta \end{pmatrix}=\begin{pmatrix} 1 & -3 \\ 3 & 3 \end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$
* $E\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix}=\begin{pmatrix} 1 & -3 \\ 3 & 3 \end{pmatrix}\begin{pmatrix} 1 \\ -1 \end{pmatrix}=\begin{pmatrix} 4 \\ 0 \end{pmatrix}$
* $\sum_{4}=\frac{1}{6}\begin{pmatrix} 1 & -3 \\ 3 & 3 \end{pmatrix}\begin{pmatrix} 5 & -2 \\ -2 & 2 \end{pmatrix}\begin{pmatrix} 1 & 3 \\ -3 & 3 \end{pmatrix}=\frac{1}{6}\begin{pmatrix} 35 & 9 \\ 9 & 27 \end{pmatrix}$
* $detZ_{4}=24$
* $D_{\xi_{4}}=\frac{35}{6}$
* $D_{\eta_{4}}=\frac{27}{6}$
* $cov\left(\xi_{4},\eta_(4)\right)=\frac{9}{6}$
* $\rho\left(\xi_{4},\eta_(4)\right)=\frac{cov\left(\xi_{4},\eta_(4)\right)}{\sqrt{D_{\xi_{4}}}*D_{\eta_{4}}}=\frac{9}{6\frac{\sqrt{35*27}}{6}}=\frac{\sqrt{105}}{35}$
* $\sum_{4}^{-1}=\frac{1}{24}\begin{pmatrix} \frac{27}{6} & -\frac{9}{6} \\ -\frac{9}{6} & \frac{35}{6} \end{pmatrix}=\begin{pmatrix} \frac{3}{16} & -\frac{1}{16} \\ -\frac{1}{16} & \frac{35}{144} \end{pmatrix}$
* $C=\frac{1}{2\pi\sqrt{det\sum_{4}}}=\frac{1}{2\pi2\sqrt{6}}=\frac{1}{4\sqrt{6}\pi}$
* $P_{\xi_{4}\eta_{4}}\left(x,y\right)=\frac{1}{4\sqrt{6}\pi}exp\left\lbrace -\frac{1}{2} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} 4 \\ 0 \end{pmatrix} \right\rbrack^{\intercal} \begin{pmatrix} \frac{3}{16} & -\frac{1}{16} \\ -\frac{1}{16} & \frac{35}{144} \end{pmatrix} \left\lbrack \begin{pmatrix} x \\ y \end{pmatrix}-\begin{pmatrix} 4 \\ 0 \end{pmatrix} \right\rbrack \right\rbrace$

```python
import plotly.graph_objects as go
import numpy as np
x, y = np.meshgrid(np.linspace(-5, 12, 100), np.linspace(-10, 10, 100))
z = (1/(np.sqrt(6)*np.pi*4))*np.exp(-1/2*((3/16)*(x-4)**2-2*(1/16)*(x-4)*y+(35/144)*y**2))

fig = go.Figure(data=go.Surface(x=x, y=y, z=z, colorscale="Hot"))  

fig.update_layout(
    title="График плотности распределения",
    scene=dict(
        xaxis=dict(title="X"),
        yaxis=dict(title="Y"),
        zaxis=dict(title="Z")
    )
)
fig.show()
```
