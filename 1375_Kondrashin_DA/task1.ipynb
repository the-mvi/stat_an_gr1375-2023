## Основные теоретические положения
  **Многомерное нормальное распределеение** — это обобщение одномерного нормального распределения. Случайный вектор, имеющий многомерное нормальное распределение, называется гауссовским вектором.



Случайный ветор $\vec{\xi}$ = ($\xi_1$, ... , $\xi_n$)$^T$ : $\Omega \rightarrow \mathbb{R}^n$ - называется **нормальным**(гауссовым), если существуют: вектор независимой стандартной нормальнной случайной величины $\vec{η} = (η_1, ... , η_n)^T $, матрица $А$ размерности $n \times m$ и вещественный вектор $B = (b_1, ... , b_n)^T$, такие что $\vec{\xi} = A \times \vec{\eta} + B$.

Обозначается как $\vec{\eta} \sim N(\mu, \Sigma) $, где $\mu = \mathbb{E}\vec{\xi} = (\mathbb{E}\xi_1, ... , \mathbb{E}\xi_n)^T$ - вектор мат.ожиданий компоненты, а $\Sigma = $ \{$ \sigma_{ij}$\}$^n_{ij=1}$ - ковариационная матрица, где
$\sigma_{ij} = cov(\xi_i, \xi_j)$.

**Плотность нормального вектора имеет вид:**

$p_{\vec{\xi}(x_1, ... , x_n)} = \frac {1}{(2\pi)^{n/2}\sqrt{det\Sigma}} exp((\frac{-1}{2}\vec{x} - \mu)^T \Sigma ^{-1}(\vec{x} - \mu)),$

$\vec{x} = (x_1, ... x_n)^T $

**Свойства многомерного нормального распределения:**
- Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенной
- Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределенными. $\mathcal{p}_{\xi|\eta=y}(x) = \frac{\mathcal{p}_{\xi,\eta}(x,y)}{\mathcal{p}_{\eta}(y)}$
- Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
- Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}\left(\overrightarrow{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, $\overrightarrow{b}$ - произвольный вектор размерности $n \times 1$, то $A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}\left(A\overrightarrow{\mu}+\overrightarrow{b}, A \Sigma A^\intercal\right)$.

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики случайного вектора. Построить график плотности распределения.

$p_{\xi,\eta}(x,y) = C\cdot exp(-\frac{1}{2}(2x^2 + 4xy + 5y^2 - 4x + 8y + 14))$:

$q(x,y) = (2x^2 + 4xy - 4x) + (5y^2 + 8y + 14) = (2x^2 + 4x(y - 1)) + (5y^2 + 8y + 14) =\\
=2(x^2 + 2x(y-1) + (y-1)^2) + (5y^2 + 8y + 14 - 2y^2 + 4y - 2) =  2(x+(y-1))^2 + 3(y^2 + 4y + 4)=\\
2(x+(y - 1))^2 + 3(y+2)^2 = 2(x-3)^2 + 4(x-3)(y+2)+2(y+2)^2+3(y+2)^2 = 2(x - 3)^2 + 2 \cdot2 (x-3)(y+2) + 5(y+2)^2
$

$\mathbb{E}\begin{pmatrix}
    \xi \\
    \eta
  \end{pmatrix}  = \begin{pmatrix}
    3 \\
    -2
  \end{pmatrix}
  $
  <br>
$\Sigma^{-1} = \begin{pmatrix}
    2 & 2 \\
    2 & 5
  \end{pmatrix}$
<br>
  $\Sigma = \frac{1}{6} \begin{pmatrix}
    5 & -2 \\
    -2 & 2
  \end{pmatrix}$
<br>
  $ cov(\xi, \eta)  = -\frac{1}{3}$
<br>
 $\mathbb{D} \xi = \frac{5}{6} $
<br>
$\mathbb{D} \eta = \frac{1}{3} $
<br>
$\rho(\xi,\eta) = \cfrac{\cfrac{1}{3}}{\sqrt{\cfrac{5}{6}} \cdot {\sqrt{\cfrac{1}{3}}}} = \cfrac{\sqrt{10}}{5}$
<br>
$C = \frac{1}{(2\pi)^{n/2} \sqrt{(det\Sigma)}} = \frac{\sqrt{6}}{2\pi} $

**Плотность:**

$p_{\xi,\eta}(x,y) =  \frac{sqrt{6}}{2\pi} \exp\left(-\frac{1}{2}
  \left[
  \begin{pmatrix}
    x \\

  \begin{pmatrix}

  \begin{pmatrix}

  \begin{pmatrix}

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(-2, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
mu = np.array([3, -2])
cov = np.array([[5/6, -1/3], [-1/3, 1/3]]) #  матрица ковариации
inv_cov = np.linalg.inv(cov) # обратная ковариационная матрица
det_cov = np.linalg.det(cov) # det  матрицы ковариации

def pdf(x, y):
    xy = np.array([x, y]) - mu
    exponent = -0.5 * np.dot(np.dot(xy, inv_cov), xy)
    return (1 / (2 * np.pi * np.sqrt(det_cov))) * np.exp(exponent)

Z = np.vectorize(pdf)(X, Y)
#график
grf = plt.figure()
A = grf.add_subplot(111, projection='3d')
A.plot_surface(X, Y, Z, cmap='viridis')
A.set_xlabel('x')
A.set_ylabel('y')
plt.show()
```
