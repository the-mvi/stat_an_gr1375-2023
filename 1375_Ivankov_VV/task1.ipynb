## Основные теоретические положения

*Определение многомерного нормального распределения*
Случайный вектор $\vec\xi=(\xi_1, \xi_2, ..., \xi_n)^\intercal: \Omega \rightarrow \mathbb{R}^n$ - называют нормальным (гауссовым) если существуют вектор независмых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1, ... , \eta_n)^\intercal$, матрица $A$ размерности $n\times m$ и вещественный вектор $B = (b_1, ... , b_n)^\intercal$, такие что: $\vec\xi=А\vec\eta+B$

- Плотность нормального вектора имеет вид:\
$p_{\vec{\xi}}(x_1, ... ,x_{n}) = \cfrac{1}{(2\pi)^{n/_{2}}\sqrt{det\Sigma}}\cdot \exp \begin{pmatrix}-\cfrac{1}{2}(\vec{x}-\mu)^{\intercal}\Sigma^{-1}(\vec{x}-\mu)\end{pmatrix}$, где $\vec x = (x_{1}, ... , x_{n})^{\intercal}$

*Свойства многомерного нормального распределения*
 - Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределеннными.
 - Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениеми.
 - Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму
 - Многомерное нормальное распределение устойчиво относительно линейных преобразований, если $\vec\xi \sim \mathcal{N}(\mu,\Sigma)$, $A$ - произвольная матрица размерности $m \times n$, а $B$ - произвольный вектор $m \times 1$ то: $A \vec\xi + B \sim \mathcal{N}(A_{\mu} + B, A\Sigma A^{\intercal})$

#### 1.1 Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.

Применим метод Лагранжа для преобразования выражения стоящего под экспонентой:\

$2x^2 - 4xy + 5y^2 + 4x - 16y + 14$\

\
Найдем характеристики:\
\
$\vec\mu = \begin{pmatrix} 1 \\ 2\end{pmatrix}$\
$\Sigma^-1 = \begin{pmatrix} 2 & -2 \\ -2 & 5\end{pmatrix}$\

#### 1.2 Построить график плотности распределения $p_{\xi,\eta}$


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import scipy.stats as sps




cov = 1/3
d_xi = 5/6
d_eta = 1/3

def show_distr(mu, sigma ,rangee, acc = 100):
    fig = plt.figure()
    dist = sps.multivariate_normal(mu,sigma)
    
    x = np.linspace(mu[0] - rangee/2, mu[0]+rangee/2, num=acc)
    y = np.linspace(mu[1] - rangee/2,mu[1]+rangee/2, num=acc)
    x,y = np.meshgrid(x,y)
    z = dist.pdf(np.dstack((x,y)))
    
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(x,y,z,cmap = cm.coolwarm)
    ax.set_title('График плотности распределения')
    plt.show()


show_distr(mu = [1, 2], sigma = [[d_xi, cov], [cov,d_eta]], rangee=6)
```
