## Основные теоретические положения
Многомерное нормальное распределение — это обобщение одномерного нормального распределения. Случайный вектор, имеющий многомерное нормальное распределение, называется гауссовским вектором.  

Случайный вектор $\vec\xi=(\xi_1, \xi_2, ..., \xi_n)^\intercal: \Omega \rightarrow \mathbb{R}^n$ - нормальный (гауссовский), если существует $\vec\xi=А\vec\eta+B$, где вектор независимой стандартной нормальнной случайной величины $\vec{\eta} = (\eta_1, ... , \eta_n)^\intercal $, матрица $А$ размерности $n \times m$ и вещественный вектор $B = (b_1, ... , b_n)^\intercal$.  
Плотность нормального вектора имеет вид: 
$\mathrm{p}_{\vec{\xi}}(x_1, ..., x_n) = \frac1{(2\pi)^{n/2} \sqrt{\det{\Sigma}}} \exp\bigg(-\frac12 (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \bigg)$,   
где: $\mu = \mathrm{E}_{\vec{\xi}} = (\mathrm{E}_{\xi_1}, ..., \mathrm{E}_{\xi_n})^\intercal$ - вектор мат. ожидания,\
$\Sigma = \{ \mathrm{cov}(\xi_i, \xi_j) \}_{i,j=1,1}^n$ - ковариационная матрица.

**Свойства многомерного нормального распределения:**  
1) Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределеннными.
2) Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениеми.
3) Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму)
4) Многомерное нормальное распределение устойчиво относительно линейных преобразований.

$ \rho(\xi, \eta) = \frac{\mathrm{cov}(\xi, \eta)}{\sqrt{\mathrm{D}(\xi) \mathrm{D}(\eta)}} - \text{Формула для нахождения коэффициента корреляции}$

$ \mathrm{C} = \frac1{(2\pi)^\frac{n}2 \sqrt{\det{\Sigma}}} - \text{Формула для нахождения константы}$

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $\mathrm{p}_{\xi,\eta}$.


$q(x, y) = 3x^2 + 3xy + 7y^2 - 12x + 19y + 37 =
 (3x^2 + 3xy - 12x) +7y^2 + 19y + 37 =
 3(x^2 + x(y-4)) + 7y^2 + 19y + 37 =
 3(x^2 + 2x\frac{y-4}2 + (\frac{y-4}2)^2 - (\frac{y-4}2)^2) + 7y^2 + 19y + 37 =   
 3(x + \frac{y-4}2)^2 - 3\left(\frac{y-4}2\right)^2 + 7y^2 + 19y + 37 = 
 3(x + \frac{y}2 -2)^2 + \frac{25y^2 + 100y + 100}4 =   
 3((x-3) + \frac12(y+2))^2 + \frac{25}4(y+2)^2 =       
 3(x-3)^2 + 3(x-3)(y+2) + 7(y+2)^2$

### Вектор математических ожиданий и обратную ковариационную матрицу.

$\mathbb{E}\begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} 3 \\ -2 \end{pmatrix}$

$\Sigma^{-1} = \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7\end{pmatrix}$

### Найдем матрицу ковариаций.

$\Sigma =  \frac{2}{75} \begin{pmatrix} 14 & -3 \\ -3 & 6\end{pmatrix}$

### Числовые характеристики случайного вектора:

$\mathrm{D}(\xi) = \frac{28}{75} \ \,  \mathrm{D}(\eta) = \frac{4}{25} - \text{ (Дисперсия для} \ \xi \ и \ \eta)$

$\mathrm{cov}(\xi, \eta) = -\frac2{25} - \text{(Ковариация)}$
 
$C = \frac{1}{2\pi\cdot\frac{2}{5\sqrt{3}}} = \frac{5\sqrt{3}}{4\pi}$

### Найдем коэффициент корреляции.

$\rho(\xi, \eta) = \frac{\mathrm{cov}(\xi, \eta)}{\sqrt{\mathrm{D}_\xi \mathrm{D}_\eta}} = \frac{-\frac2{25}}{\sqrt{\frac{28}{75} \cdot \frac4{25}}} = -\frac{\sqrt3}{2\sqrt7}$

### Выпишем функцию плотности распределения.

$\mathcal{p}_{\xi,\eta}(x,y) = \frac{5\sqrt{3}}{4\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ -2 \end{pmatrix} \right]^\intercal \begin{pmatrix} 3 & \frac{3}{2} \\ \frac{3}{2} & 7\end{pmatrix}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 3 \\ -2 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal


# Заданные параметры
cov = np.array([[28/75, -2/25], [-2/25, 4/25]]) # матрица ковариаций
vector = np.array([3,-2]) # вектор математических ожиданий
random = 11

#Создаем распределения
distribution = multivariate_normal(cov = cov, mean = vector, seed = random)

# Создаем сетку 
x = np.linspace(vector[0] - 3*np.sqrt(cov[0,0]), vector[0] + 3*np.sqrt(cov[0,0]), num=2023)
y = np.linspace(vector[1] - 3*np.sqrt(cov[1,1]), vector[1] + 3*np.sqrt(cov[1,1]), num=2023)
X, Y = np.meshgrid(x,y)
pos = np.dstack((X, Y))

# Вычисляем значения плотности вероятности 
density = distribution.pdf(pos)

# Создаем трехмерный график
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, density, cmap = "plasma")

ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')
plt.title("График плотности вероятности случайного вектора")
plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(- 2\xi - 5\eta, \xi + 2\eta)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.


Выпишем матрицу линейного преобразования.

$A = \begin{pmatrix} -2 & -5 \\ 1 & 2\end{pmatrix}$

Т.к. $\det A = 1$, то матрица невырожденная . Вычислим характеристики распределения.

$\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -2 & -5 \\ 1 & 2\end{pmatrix}\begin{pmatrix} \xi \\ \eta \end{pmatrix}$

$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -2 & -5 \\ 1 & 2\end{pmatrix} \begin{pmatrix} 3 \\ -2 \end{pmatrix} = \begin{pmatrix} 4 \\ -1 \end{pmatrix}$

$\Sigma_{4} = \frac{2}{75}\begin{pmatrix} -2 & -5 \\ 1 & 2\end{pmatrix}\begin{pmatrix} 14 & -3 \\ -3 & 6\end{pmatrix}\begin{pmatrix} -2 & 1 \\ -5 & 2\end{pmatrix} = \frac{2}{75}\begin{pmatrix} -13 & -24 \\ 8 & 9\end{pmatrix}\begin{pmatrix} -2 & -5 \\ 1 & 2\end{pmatrix} = \frac{2}{75}\begin{pmatrix} 146 & -61 \\ -61 & 26\end{pmatrix} $

$\mathbb{D}\xi_{4} = \frac{292}{75},\ \mathbb{D}\eta_{4} = \frac{52}{75},\ \mathop{\mathrm{cov}}(\xi_{4}, \eta_{4}) = -\frac{122}{75}$


$\rho(\xi_4, \eta_4) = \frac{\mathrm{cov}(\xi_4, \eta_4)}{\sqrt{\mathrm{D}_{\xi_4}\mathrm{D}_{\eta_4}}} = \frac{-\frac{122}{75}}{\sqrt{\frac1{75} \cdot 292 \cdot \frac4{75} \cdot 13}} = \frac{-61}{\sqrt{292 \cdot 13}}$   

$  \Sigma_4^{-1}=  \frac2{75} \left( \begin{matrix} 26 & 61\\ 61 & 146 \end{matrix} \right) $

$C = \frac{1}{2\pi\cdot\frac{16}{5\sqrt{3}}} = \frac{5\sqrt{3}}{32\pi}$

Выпишем плотность полученного распределения.

$\mathcal{p}_{\xi_4,\eta_4}(x,y) = \frac{5\sqrt{3}}{32\pi} \exp\left\{-\frac{1}{2}\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 4 \\ -1 \end{pmatrix} \right]^\intercal \frac2{75} \left( \begin{matrix} 26 & 61\\ 61 & 146 \end{matrix} \right)\left[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} 4 \\ -1 \end{pmatrix} \right]\right\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal


# Заданные параметры
cov = np.array([[292/75, -122/75], [-122/75, 52/75]]) # матрица ковариаций
vector = np.array([4, -1]) # вектор математических ожиданий
random = 11

#Создаем распределение
distribution = multivariate_normal(cov = cov, mean = vector, seed = random)

# Создаем сетку 
x = np.linspace(vector[0] - 3*np.sqrt(cov[0,0]), vector[0] + 3*np.sqrt(cov[0,0]), num=2023)
y = np.linspace(vector[1] - 3*np.sqrt(cov[1,1]), vector[1] + 3*np.sqrt(cov[1,1]), num=2023)
X, Y = np.meshgrid(x,y)
pos = np.dstack((X, Y))

# Вычисляем значения плотности вероятности 
density = distribution.pdf(pos)

# Создаем трехмерный график
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, density, cmap = "plasma")

ax.set_xlabel('ось X')
ax.set_ylabel('ось Y')
plt.title("График плотности вероятности вектора с независимыми компонентами")
plt.show()
```
