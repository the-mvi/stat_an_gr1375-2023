## Основные теоретические положения
1. ***Случайным вектором*** $\overrightarrow{\xi}=(\xi_1,...,\xi_n)^\mathsf{T} : Ω \rightarrow R^n$ называют *нормальным(гауссовым)*, если существует вектор независимых стандартных нормальных величин $\overrightarrow{\eta}=(\eta_1,...,\eta_n)^\mathsf{T}$, матрица A размерности $n\times m$ и вектор $B=(b_1,...,b_m)^\mathsf{T} $, такие что:
  * $\overrightarrow{\xi}=\overrightarrow{A_\eta}+B$
  * $\overrightarrow{\xi} \sim \mathcal{N}(\mu,\Sigma)$ , где $ \mu = \mathbb{E} \overrightarrow{ξ} = (\mathbb{E} \xi_1,...,\mathbb{E} \xi_n)^\mathsf{T}$ - вектор математических ожиданий компонент, $\Sigma = \{ \sigma_\overrightarrow{\eta}\}^n _{i,j=1} $ , $\sigma_{ig} = cov(\xi_i,\xi_j)$ - ковариационная матрица.

2. ***Плотность нормального вектора*** имеет вид: $ P_\overrightarrow{\xi}(x_1,...,x_n) = \frac{1}{(2\pi)\sqrt{det\Sigma}}exp(-\frac{1}{2}(\overrightarrow{x}-\mu)^\mathsf{T}\Sigma^{-1}(\overrightarrow{x}-\mu))$

3. ***Коэффициент корреляции*** - это мера линейной зависимости двух случайных величин: $p(\xi,\eta) = \frac{cov(\xi,\eta)}{\sqrt{\mathbb{D}(\xi)\mathbb{D}(\eta)}}$,
где $cov(x,y)$ - ковариация, а $\mathbb{D}$ - дисперсия

4. ***Ковариационная матрица***
$\Sigma =
\begin{pmatrix}
\mathbb{D}\xi & cov(\xi,\eta) \\
cov(\eta,\xi) &  \mathbb{D}\eta
\end{pmatrix} $

5. ***Свойства многомерного нормального распределения:***
  * Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределёнными.
  * Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределёнными.
  * Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
  * Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\overrightarrow{\xi}\sim \mathcal{N}(\mu,\sigma),A $ - произвольная матрица размерности $m\times n$, а $B$ - произвольный вектор размерности $m\times 1$, то $A\overrightarrow{\xi} + B \sim \mathcal{N}(A\mu,A\Sigma A^\mathsf{T})$

###1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $ \rho_{\xi,\eta}$
* Приведение квадратичной формы к каноническому виду методом Лагранжа: $q(x,y) = 2x^2 + 4xy +5y^2 +12x+ 18y+21 = 2\cdot((x^2+2x\cdot(y+3)+(y+3)^2)-2\cdot(y+3)^2+5y^2+18y+21) = 2\cdot(x+y+3)^2+3\cdot(y+1)^2=2\cdot((x+2)+(y+1))^2+3\cdot(y+1)^2=2\cdot(x+2)^2+2\cdot2\cdot(x+2)\cdot(y+1)+5\cdot(y+1)^2$
* Вектор математического ожидания:

 $\mathbb{E}\begin{pmatrix}

* Обратная ковариационная матрица:

 $\Sigma^{-1}=\begin{pmatrix}
2 & 2 \\

* Ковариационная матрица:

 $\Sigma= \frac{1}{det\Sigma^{-1}}
\begin{pmatrix}

\begin{pmatrix}

 $\rho(\xi,\eta)=\frac{cov(\xi,\eta)}{\sqrt{\mathbb{D}(\xi)\mathbb{D}(\eta)}}=-\frac{\sqrt{10}}{5}$
* Коэффициент C:

 $det\Sigma = \frac{1}{6}$

 $C = \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{det\Sigma}} = \frac{\sqrt{6}}{2\pi}$

* Плотность:

 $p_{\xi,\eta}(x,y) = \frac{\sqrt{6}}{2\pi} \cdot \exp \left\{ - \frac{1}{2}

```python
import matplotlib.pyplot as plt
import numpy as np

def p(x,y):
  return (6**(1/2))/(2*np.pi)*np.exp(-1/2*(2*x**2+4*x*y+5*y**2+12*x+18*y+21))

E1 = -2
E2 = -1

x = np.linspace(E1-3,E1+3,100)
y = np.linspace(E2-3,E2+3,100)
X,Y = np.meshgrid(x,y)
p = p(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_title('График плотности распределения', fontsize=16)
ax.plot_surface(X, Y, p, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('p')

plt.show()
```
