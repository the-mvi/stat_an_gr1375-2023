## Основные теоретические положения
Случайный вектор $\overrightarrow\xi=\left(\xi_1,...,\xi_n\right)^{\intercal}:\;\Omega\rightarrow R^n$ называют нормальным (гауссовым) если существует вектор назависимых стандартных нормальных случайных величин $\overrightarrow\eta=\left(\eta_1,...,\eta_m\right)^{\intercal}$, матрица A размерности $n\times m$ и вектор $B=\left(b_1,...,b_m\right)^\intercal$, такие что:<br>
$\overrightarrow\xi=A\overrightarrow\eta+B$<br>
$\overrightarrow\xi\sim \mathcal{N}(\mu,\Sigma)$ где $\mu=\mathbb{E}\overrightarrow\xi=\left(\mathbb{E}\xi_1,...,\mathbb{E}\xi_n\right)^{\intercal}$ - вектор математических ожиданий компонент, $\Sigma={\left\{\sigma_\overrightarrow\eta\right\}^n}_{i,j=1}$, $\sigma_{ij}=\mathrm{cov}\left(\xi_i,\xi_j\right)$ - ковариационная матрица<br><br>
Плотность нормального вектора имеет вид: $ P_{\overrightarrow{\xi}}(x_{1},...,x_{n}) = \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{\det\Sigma}}\exp\left(-\frac{1}{2}\left(\overrightarrow{x}-\mu\right)^{\intercal}\Sigma^{-1}\left(\overrightarrow{x}-\mu\right)\right)$<br><br>
Свойства:<br>
1. Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределенными.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями. $P_{(\xi\vert\eta=y)}=\frac{P_{\xi,\eta}(x,y)}{P_\eta(y)}$
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований. Если $\overrightarrow{\xi} \sim \mathcal{N}(\overrightarrow{\mu}, \Sigma)$, А произвольная матрица размерности $m \times n, \overrightarrow{b}$ произвольный вектор размерности $m \times 1$, то $ A\overrightarrow{\xi}+\overrightarrow{b} \sim \mathcal{N}(A\overrightarrow{\mu}, A\Sigma A^{\intercal})$

<h3>Задание 1</h3>Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $P_{\xi,\eta}$<br><br>
$q(x,y)=3x^2-3xy+7y^2-9x-8y+13=\left(3x^2-3xy-9x\right)+7y^2-8y+13=\left(3x^2-3x\left(y+3\right)\right)+7y^2-8y+13=3\left(x^2-x\left(y+3\right)+\frac{\left(y+3\right)^2}4\right)-\frac{3\left(y+3\right)^2}4+7y^2-8y+13=3\left(x-\frac{y+3}2\right)^2-\frac{3y^2+18y+27}4+7y^2-8y+13=3\left(x-\frac12y-\frac32\right)^2+\frac{25}4y^2-\frac{50}4y+\frac{25}4=3\left(x-\frac12y-\frac32\right)^2+\frac{25}4\left(y^2-2y+1\right)=3\left(x-\frac12y-\frac32\right)^2+\frac{25}4\left(y-1\right)^2=3\left(\left(x-2\right)-\frac12\left(y-1\right)\right)^2+\frac{25}4\left(y-1\right)^2=3\left(x-2\right)^2-3\left(x-2\right)\left(y-1\right)+\frac34\left(y-1\right)^2+\frac{25}4\left(y-1\right)^2=3\left(x-2\right)^2-3\left(x-2\right)\left(y-1\right)+7\left(y-1\right)^2$

Вектор математического ожидания<br><br>
$\mathbb{E}\begin{pmatrix}\xi\\\eta\end{pmatrix}=\begin{pmatrix}2\\1\end{pmatrix}$<br><br>
Ковариационная матрица<br><br>
$\Sigma^{-1}=\begin{pmatrix}3&-1.5\\-1.5&7\end{pmatrix}\\\Sigma=\begin{pmatrix}3&-1.5\\-1.5&7\end{pmatrix}^{-1}=\frac4{75}\begin{pmatrix}7&1.5\\1.5&3\end{pmatrix}$
<br>
$\det\;\Sigma\;=\;7\cdot3-1.5\cdot1.5=18.75=\frac{75}4\\\mathbb{D}\left(\xi\right)=\frac{\displaystyle7}{\displaystyle\frac{75}4}=7\cdot\frac4{75}=\frac{28}{75}$
<br>Дисперсия, ковариация, корреляция<br><br>
$\mathbb{D}\left(\eta\right)=\frac3{\displaystyle\frac{75}4}=3\cdot\frac4{75}=\frac{12}{75}=\frac4{25}\\\mathrm{cov}\left(\xi,\eta\right)=\frac{1.5}{\displaystyle\frac{75}4}=\frac32\cdot\frac4{75}=\frac2{25}\\\rho\left(\xi,\eta\right)=\frac{\mathrm{cov}\left(\xi,\eta\right)}{\sqrt{\mathbb{D}\left(\xi\right)\cdot \mathbb{D}\left(\eta\right)}}=\frac{{\displaystyle\frac2{25}}}{\sqrt{{\displaystyle\frac{28}{75}}\cdot{\displaystyle\frac4{25}}}}=\frac{\sqrt{21}}{14}$

<br>Плотность нормального распределения<br><br>
$C=\frac1{\left(2\pi\right)^\frac n2\cdot \sqrt{\det\Sigma}}=\frac1{2\pi\cdot\frac2{5\sqrt{3}}}=\frac{5\sqrt{3}}{4\pi}\\P_{\left(\xi,\eta\right)}=\frac{5\sqrt3}{4\pi}\exp{\left(-\frac12\left(\left[\begin{pmatrix}x\\y\end{pmatrix}-\begin{pmatrix}2\\1\end{pmatrix}\right]^{\intercal}\begin{pmatrix}3&-1.5\\-1.5&7\end{pmatrix}\left[\begin{pmatrix}x\\y\end{pmatrix}-\begin{pmatrix}2\\1\end{pmatrix}\right]\right)\right)}$

```python
# График функции плотности распределения
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def density(x, y):
    return (5*np.sqrt(3)/(4*np.pi))*np.exp(-1/2*(3*(x-2)**2-3*(x-2)*(y-1)+7*(y-1)**2))

x = np.linspace(-2, 5, 50)
y = np.linspace(-2, 5, 50)
X, Y = np.meshgrid(x, y)
P=density(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, P, cmap='inferno')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('P')

plt.show()
```

<h3>Задание 3</h3>Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.<br><br>

$\Sigma_3=Q^{\intercal}\Sigma Q=\frac4{150}\begin{pmatrix}3&1\\1&-3\end{pmatrix}\begin{pmatrix}7&1.5\\1.5&3\end{pmatrix}\begin{pmatrix}3&1\\1&-3\end{pmatrix}=\begin{pmatrix}\frac25&0\\0&\frac2{15}\end{pmatrix}$<br>

$P_{\xi_3,\eta_3}=\frac{5\sqrt3}{4\pi}\exp{\left(-\frac12\left(\frac12\left(5{(x-\frac7{\sqrt{10}})}^2+15{(y+\frac1{\sqrt{10}})}^2\right)\right)\right)}$<br><br>

```python
# График функции плотности распределения вектора с независимыми компонентами
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def density(x, y):
    return (5*np.sqrt(3)/(4*np.pi))*np.exp(-1/4*(5*(x-7/np.sqrt(10))**2+15*(y+1/np.sqrt(10))**2))

x = np.linspace(-2, 5, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)
P=density(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, P, cmap='plasma')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('P')

plt.show()
```

<h3>Задание 4</h3>Вычислить характеристики совместного распределения случайного вектора $(-4\xi + 2\eta, \xi + 2\eta)^{\intercal}$ и записать его плотность. Построить график плотности полученного распределения.<br><br>
$A=\begin{pmatrix}-4&2\\1&2\end{pmatrix}$<br><br>Вектор математического ожидания<br><br>
$\mathbb{E}\begin{pmatrix}\xi_4\\\eta_4\end{pmatrix}=A\mathbb{E}\begin{pmatrix}\xi\\\eta\end{pmatrix}=\begin{pmatrix}-4&2\\1&2\end{pmatrix}\begin{pmatrix}2\\1\end{pmatrix}=\begin{pmatrix}-6\\4\end{pmatrix}$<br><br>Ковариационная матрица<br><br>
$\Sigma_4=A\Sigma A^{\intercal}=\frac4{75}\begin{pmatrix}-4&2\\1&2\end{pmatrix}\begin{pmatrix}7&1.5\\1.5&3\end{pmatrix}\begin{pmatrix}-4&1\\2&2\end{pmatrix}=\frac13\begin{pmatrix}16&-4\\-4&4\end{pmatrix}$<br><br>
$\det\Sigma_4=\frac{16}3\cdot\frac43-\frac43\cdot\frac43=\frac{16}3$<br><br>
$\Sigma_4^{-1}=\frac3{16}\begin{pmatrix}\frac43&\frac43\\\frac43&\frac{16}3\end{pmatrix}=\frac14\begin{pmatrix}1&1\\1&4\end{pmatrix}$<br><br>Дисперсия, ковариация, корреляция<br><br>
$\mathbb{D}_\xi=\frac{16}{3};\;\;\;\mathbb{D}_\eta=\frac{4}{3};\;\;\;\mathrm{cov}\left(\xi,\eta\right)=-\frac{4}{3}$<br><br>
$\rho\left(\xi,\eta\right)=\frac{\mathrm{cov}\left(\xi,\eta\right)}{\sqrt{\mathbb{D}\left(\xi\right)\cdot \mathbb{D}\left(\eta\right)}}=-\frac4{3\sqrt{\displaystyle\frac{64}{9}}}=-\frac12$<br><br>Плотность совместного распределения случайного вектора<br><br>
$C_4=\frac1{\left(2\pi\right)^\frac n2\cdot \sqrt{\det\Sigma_4}}=\frac1{2\pi\sqrt{\displaystyle\frac{16}3}}=\frac{\sqrt3}{8\pi}$<br><br>
$q(x,y)=\frac14\left(x+6\right)^2+\frac12\left(x+6\right)\left(y-4\right)+\left(y-4\right)^2$<br><br>
$P_{\xi_4,\eta_4}=\frac{\sqrt3}{8\pi}\exp{\left(-\frac12\left(\frac14\left(x+6\right)^2+\frac12\left(x+6\right)\left(y-4\right)+\left(y-4\right)^2\right)\right)}$<br><br>
$P_{\left(\xi_4,\eta_4\right)}=\frac{\sqrt3}{8\pi}\exp\left(-\frac12\left(\left[\begin{pmatrix}x\\y\end{pmatrix}-\begin{pmatrix}-6\\4\end{pmatrix}\right]^\intercal\begin{pmatrix}0.25&0.25\\0.25&1\end{pmatrix}\left[\begin{pmatrix}x\\y\end{pmatrix}-\begin{pmatrix}-6\\4\end{pmatrix}\right]\right)\right)$

```python
# График функции плотности совместного распределения случайного вектора
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def density(x, y):
    return (np.sqrt(3)/(8*np.pi))*np.exp(-1/2*(0.25*(x+6)**2+0.5*(x+6)*(y-4)+(y-4)**2))

x = np.linspace(-15, 2, 50)
y = np.linspace(0, 8, 50)
X, Y = np.meshgrid(x, y)
P=density(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, P, cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('P')

plt.show()
```
