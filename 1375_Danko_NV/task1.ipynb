## Основные теоретические положения


Случайный вектор $\overrightarrow{\xi} = \left(\xi_1,...,\xi_n\right)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют **нормальным**(гауссовым) если существует вектор независимых стандартных нормальных случайных величин $\overrightarrow{\eta} = \left(\eta_1,...,\eta_m\right)^\intercal$, матрица A размерности $m \times n$ и вещественный вектор $\mathbf{B} = \begin{pmatrix} b_1, \dotsc, b_m \end{pmatrix}^\intercal$, такие что:

$$\overrightarrow{\xi} = \mathrm{A} \overrightarrow{\eta}+ \mathrm{B}$$

**Плотность нормального вектора** имеет вид: 

$p_{\overrightarrow{\xi}}\left(x_1,...,x_n\right) = \frac{1}{{\left(2\pi\right)^\frac{n}{2}\sqrt{\begin{vmatrix} \Sigma \end{vmatrix}}}}\mathrm{exp}\left(\frac{1}{2}\left(\overrightarrow{x} - \mu \right)^\intercal \Sigma^{-1}\left(\overrightarrow{x} - \mu \right)\right)$


**Свойства многомерного нормального распределения:**


*   Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными.
*   Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальнымы распределениями.
*   Параметры многомерного нормального распледеления(среднее и ковариационная матрица) полностью описывают его форму.
*   Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\overrightarrow{\xi} \sim \mathcal{N}\left(\mu,\Sigma\right)$, $\mathrm{A}$ - произвольная матрица размерности $m \times n$, а $\mathrm{B}$ - произвольный вектор размерности  $m \times 1$, то
$$\mathrm{A}\overrightarrow{\xi} + \mathrm{B} \sim \mathcal{N}\left(\mathrm{A}\mu + \mathrm{B},\mathrm{A}\Sigma\mathrm{A}^\intercal\right) $$

Формула для плотности условного распределения:  

$p_{\xi|\eta=y}(x) = \frac{p_{\xi, \eta}(x, y)}{p_{\eta}(y)}$

Коэффициент корреляции - это мера линейной зависимости двух случайных величин:
$$\rho(x,y) = \frac{\mathrm{cov}(x, y)}{\sqrt{\mathbb{D}x\cdot\mathbb{D}y}}$$
где $\mathrm{cov}(x,y)$ - ковариация, а $\mathbb{D}$-дисперсия

###1. Вычислить вектор мат. ожиданий и ковариационные характеристики случайного вектора. Построить график плотности распределения.


$p_{\xi,\eta}(x,y) = \mathrm{C}\cdot \mathrm{exp}(\frac{-1}{2}(4x^2 + 4xy+7y^2+16x-16y+40))$

Приведение квадратурной формы к каноническому виду:
$q(x,y) = 4x^2 + 4xy+7y^2+16x-16y+40 =$

$ 4(x^2+xy+4x) + 7y^2-16y+40 $


$ 4(x^2+x(y+4))+7y^2 -16y +40 $

$ 4(x^2 + 2x\frac{x(y+4)}{2} + (\frac{y+4}{2})^2 - (\frac{y+4}{2})^2) + 7y^2-16y+40$

$ 4(x + \frac{y+4}{2})^2 - (y^2+8y+16) + 7y^2-16y+40$  

$ 4(x + \frac{y+4}{2})^2 - y^2 -8y -16 +7y^2 -16y +40$  

$ 4(x+y/2+2)^2 +6(y^2-4y+4)$  

$ (2x+y+4)^2 + 6(y-2)^2 $

$ (2x+6+(y-2))^2 +6(y-2)^2$

$ (2x+6)^2 + 2(2x+6)(y-2) + 7(y-2)^2$

$ 4(x+3)^2 +2*2(x+3)(y-2)+7(y-2)^2$ 

 
Вектор мат. ожидания:

$\mathbb{E}\begin{pmatrix}
    \xi \\
    \eta
  \end{pmatrix}  = \begin{pmatrix}
    -3 \\
    2
  \end{pmatrix}$

Обратная ковариационная матрица:

$\Sigma^{-1} = \begin{pmatrix}
  4 & 2\\ 2 & 7
  \end{pmatrix}
$

Поиск ковариационной матрицы:

$\mathrm{det}\Sigma^{-1} = 28-4=24$

$\Sigma = \frac{1}{det\Sigma^{-1}}\cdot\Sigma^{-1} = \frac{1}{24}\begin{pmatrix}
 7 & -2 \\ -2 &4
 \end{pmatrix}$

Дисперсия $\xi$ и $\eta$:

$\Sigma=\begin{pmatrix} \mathbb{D}_\xi & \mathrm{cov}(\xi{,}\eta)\\ \mathrm{cov}(\eta{,}\xi) & \mathbb{D}_\eta \end{pmatrix}$

$\mathbb{D}(\xi) = \frac{7}{24}$

$\mathbb{D}(\eta) = \frac{4}{24}$

Коэффициент корреляции:

$\mathrm{cov}(\xi,\eta) = \frac{-2}{24}$

$\rho(\xi,\eta) = \frac{\mathrm{cov}(\xi,\eta)}{\sqrt{\mathbb{D}(\xi)\cdot \mathbb{D}(\eta)}} = \frac{-\frac{1}{12}}{\sqrt{\frac{7}{24}\cdot\frac{4}{27}}} = -\frac{3\sqrt{14}}{28}$

Коэффициент $C$:

$C=\cfrac{1}{2\pi\sqrt{det\Sigma}}$

$\mathrm{C} = \cfrac{1}{2\pi\sqrt{\cfrac1{24}}}=\cfrac{\sqrt{6}}{\pi}$ 

Плотность распределения:

$p_{\xi,\eta}(x,y)=\cfrac{\sqrt{6}}{\pi}\cdot\exp\Bigl\{-\cfrac{1}{2}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} -3 \\ 2\end{pmatrix}\biggr]^{\intercal}\begin{pmatrix} 4 & 2\\ 2 & 7 \end{pmatrix}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} -3 \\ 2\end{pmatrix}\biggr]\Bigr\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

C = np.sqrt(6)/np.pi
mean_vector = np.array([-3, 2]) 
cov_matrix = np.array([[7/24, -2/24], [-2/24, 4/24]])  
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.zeros(X.shape)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        point = np.array([X[i, j], Y[i, j]])
        deviation = point - mean_vector
        exponent = np.exp(-0.5 * np.dot(np.dot(deviation.T, np.linalg.inv(cov_matrix)), deviation))
        density = C * exponent
        Z[i, j] = density
fig = plt.figure(figsize=(10, 9))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='jet')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('График плотности двумерного нормального распределения')
plt.show()
```

###3. Нахождение ортогонального преобразования, переводящего соответсвующий центрированный случайный вектор в вектор с независимыми компонентами. Построение графика плотности полученного распределения.

$\Sigma^{-1} = \begin{pmatrix}4 &2\\ 2 & 7  \end{pmatrix}$

$\Sigma_{3}=\cfrac{1}{24}\cdot\cfrac{1}{5}\begin{pmatrix} -2&1 \\ 1&2\end{pmatrix} \begin{pmatrix}4&2 \\ 2 &7\end{pmatrix}\begin{pmatrix} -2&1 \\ 1&2\end{pmatrix}=\begin{pmatrix} \frac{1}{8}& 0 \\ 0 &\frac{1}{3}\end{pmatrix}$

###4. Вычислить характеристики совместного распределения случайного ветора $(\xi -3\eta, 3\xi -3\eta)^\intercal$ и записать его плотность. Построить график плотности полученного рапределения.




$\mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4}\end{pmatrix}=\begin{pmatrix} 1&-3 \\ 3&-3\end{pmatrix}\begin{pmatrix} \xi \\ \eta\end{pmatrix}=\begin{pmatrix} 1&-3 \\ 3&-3\end{pmatrix}\begin{pmatrix} -3 \\ 2\end{pmatrix}=\begin{pmatrix} -9 \\-15\end{pmatrix}$  
$\Sigma_{4}=A\cdot\Sigma\cdot A^{\intercal}=\cfrac{1}{24}\cdot\begin{pmatrix} 1&-3 \\ 3&-3\end{pmatrix}\begin{pmatrix}7&-2 \\ -2&4\end{pmatrix}\begin{pmatrix} 1&3 \\ -3&-3\end{pmatrix}=\begin{pmatrix} \cfrac{55}{24}&\cfrac{27}{8} \\ \cfrac{27}{8}&\cfrac{45}{8}\end{pmatrix}$   
$\mathrm{cov}(\xi,\eta)=\cfrac{27}{8}$  

$\mathbb{D}_{\xi_4}=\cfrac{55}{24}$    $\mathbb{D}_{\eta_4}=\cfrac{45}{8}$  

$\rho(\xi_4,\eta_4)=\frac{\cfrac{27}{8}}{\sqrt{\cfrac{55}{24}}\cdot\sqrt{\cfrac{45}{8}}}=\cfrac{9\sqrt{33}}{55}$  

$C=\cfrac{1}{2\pi\sqrt{det\Sigma_4}}=\cfrac{1}{2\pi\sqrt{\cfrac{3}{2}}}=\cfrac{\sqrt{6}}{6\pi}$  

$p_{\xi_4,\eta_4}(x,y)=\cfrac{\sqrt{6}}{6\pi}\cdot\exp\Bigl\{-\cfrac{1}{2}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} -9 \\ -15\end{pmatrix}\biggr]^{\intercal}\begin{pmatrix} \cfrac{15}{4} & -\cfrac{9}{4}\\ -\cfrac{9}{4} & \cfrac{55}{36} \end{pmatrix}\biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} -9 \\ -15\end{pmatrix}\biggr]\Bigr\}$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

cov = np.array([[7/24, -2/24], [-2/24, 4/24]]) 
mean = np.array([-3, 2])
distr = multivariate_normal(mean, cov)
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
Z = np.zeros((100, 100))
for i, xi in enumerate(x):
    for j, yj in enumerate(y):
        Z[i, j] = distr.pdf([xi, yj])
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(x, y)
ax.plot_surface(X, Y, Z, cmap='hot')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('График плотности совместного распределения')
plt.show()
```
