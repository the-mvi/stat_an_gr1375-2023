## Основные теоретические положения

Случайный вектор $\vec{\xi}=\left(\xi_1, ..., \xi_n\right)^{\mathsf{T}}: \Omega\rightarrow \mathbb{R}^n$ называют нормальным, если существует вектор стандартных случайных величин $\vec{\eta} = \left(\eta_1, ...,\eta_m\right)^{\mathsf{T}}$,<br>
матрица А размером $(n\times m)$ и вещественный $\vec{B} = \left(b_1, ...b_n\right)^{\mathsf{T}}:$
$\vec{\xi} = A\vec{\eta} + \vec{B}$

Плотность нормального вектора:
$$p_{\vec{\xi}}(\vec{x}) = \frac{1}{(2\pi)^{(n/2)}\det{\Sigma}} \cdot \exp{\left(-\frac{1}{2}(\vec{x} - \vec{\mu})^{\mathsf{T}}\Sigma^{-1}(\vec{x} -\vec{\mu})\right)}$$
$$\vec{\xi} \sim \mathcal{N}(\mu, \Sigma)$$
$$\vec{\mu} = \mathbb{E}\vec{\xi}=(\mathbb{E}\xi_1, ... ,\mathbb{E}\xi_n)^{\mathsf{T}}$$
$$\Sigma = \{\sigma_{i,j}\}^n_{i, j=1,1}, \quad \sigma_{i, j} = \mathop{\mathrm{cov}}\nolimits(\xi_i, \xi_j)$$
Свойства:
1. Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределёнными

2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распрделёнными

3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму

4. Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi}\sim\mathcal{N}(\mu, \Sigma)$, A - произвольная матрица размерности $m\times n$, а B - произвольный вектор размерности $m\times 1$, то: $A\vec{\xi} + B \sim \mathcal{N}(A\mu+B,A\Sigma A^{\mathsf{T}})$

Формула плотности условного распределения:
* $\large p_{\xi|\eta} = \frac{p_{\xi, \eta}(x, y)}{p_{\eta}(y)}$,

##### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. <br> Построить график плотности распределения $p_{\xi,\eta}$.

* Преобразуем выражение методом Лагранжа:<br><br>
    $
    q(x, y) = 3x^2 + 4xy +6y^2 - 8x + 4y + 10 = 12 = 3\left(x^2 + \frac{4}{3}xy - \frac{8}{3}x\right) + 6y^2 + 4y + 10 =\\
    = 3\left(x^2 + 2x\left(\frac{2}{3}y - \frac{4}{3}\right) + \left(\frac{2}{3}y - \frac{4}{3}\right)^2\right) + 6y^2 + 4y + 10 -

    = \textcolor{cyan}{3\left(x + \frac{2}{3}y - \frac{4}{3}\right)^2 + \frac{14}{3}(y + 1)^2} = 3\left((x - 2) + \frac{2}{3}(y + 1)\right)^2 + \frac{14}{3}(y + 1)^2 =\\
    = 3\left((x - 2)^2 + \frac{4}{3}(y + 1)(x - 2) + \frac{4}{9}(y + 1)^2\right) + \frac{14}{3}(y + 1)^2 =\\
    = 3(x - 2)^2 + 4(y + 1)(x - 2) + 6(y + 1)^2
    $

* Вектор математического ожидания:<br><br>
$\mathbb{E}\begin{pmatrix}\xi\\\eta\end{pmatrix}=\begin{pmatrix}2\\-1\end{pmatrix}$

* Обратная ковариационная и ковариационная матрицы:<br><br>
$\Sigma^{-1} = \begin{pmatrix}3&2\\2&6\end{pmatrix}$<br><br>
$\Sigma = \frac{1}{\mathrm{det}\left(\Sigma^{-1}\right)}\begin{pmatrix}\frac{18}{3}&-2\\-2&3\end{pmatrix} =

$\mathrm{cov}(\xi,\eta)=-\frac{1}{7}$<br><br>
$\rho(\xi,\eta)=\frac{\mathrm{cov}(\xi,\eta)}{\sqrt{\mathbb{D}(\xi)\cdot\mathbb{D}(\eta)}}=-\frac{\sqrt{2}}{3}$<br><br>
$C=\frac{1}{(2\pi)^{\frac{n}{2}}\cdot\sqrt{\mathrm{det}\Sigma}}=\frac{\sqrt{14}}{2\pi}$<br><br>

* Плотность распределения:<br><br>
$$p_{\xi,\eta}(x,y)=\frac{\sqrt{14}}{2\pi}\cdot\mathrm{exp}\left(-\frac{1}{2}\cdot\left[\begin{pmatrix}x\\ y\end{pmatrix}-\begin{pmatrix}2\\-1\end{pmatrix}\right]^{\mathsf{T}}\cdot\begin{pmatrix}3&2\\2&\frac{18}{3}\end{pmatrix}\cdot\left[\begin{pmatrix}x\\ y\end{pmatrix}-\begin{pmatrix}2\\-1\end{pmatrix}\right]\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_density(density_func, mean, eps = 2):
    x = np.linspace(mean[0] - eps, mean[0] + eps, num = 100)
    y = np.linspace(mean[1] - eps, mean[1] + eps, num = 100)
    X, Y = np.meshgrid(x, y)
    p = density_func(X, Y)

    ax = plt.axes(projection="3d")
    ax.set_title("Плотность распределения", fontsize=24)
    ax.plot_surface(X, Y, p, cmap="magma")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_zlabel("p")

    plt.show()


density_func = lambda x,y: (14 ** 0.5 / (2 * np.pi)) * np.exp(-(1/2)*(3*x**2+4*x*y+6*y**2-8*x+4*y+10))
mean = np.array([2, -1])

plot_density(density_func, mean)
```
