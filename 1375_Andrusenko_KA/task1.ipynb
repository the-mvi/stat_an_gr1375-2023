## Основные теоретические положения
Случайный вектор $\vec{\xi} = (\xi_1, ..., \xi_n)^\intercal : \Omega \rightarrow \mathbb{R}^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1, ..., \eta_n)^\intercal$, матрица $A$ размерности $n \times m$ и вещественный вектор $\vec{b} = (b_1, ..., b_n)^\intercal$, такие что $\vec{\xi} = A\vec{\eta} + \vec{b}$
Обозначение: $\vec{\xi} \sim \mathcal{N}(\mu, \Sigma)$
где $ \mu = \mathbb{E}\vec{\xi}=(\mathbb\xi_1,...,\mathbb\xi_n)^{\intercal}$ - вектор математических ожиданий компонент; $\Sigma = \{ \mathrm{cov}(\xi_i, \xi_j) \}_{i,j=1,1}^n $ - ковариационная матрица.

Плотность распределения нормального случайного вектора имеет вид:

$p_{\vec{\xi}}(x_1, ..., x_n) = \frac1{(2\pi)^{\frac{n}2} \sqrt{\det{\Sigma}}} \exp\left(-\frac12 (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \right)$,

где $ \vec{x}=(x_1,...,x_n)^{\intercal} $

Свойства многомерного нормального распределения:
1. Линейная комбинация случайных переменных, принадлежащих многомерному нормальному распределению, также является нормально распределённой.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормально распределёнными.
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}\left(\vec{\mu}, \Sigma\right)$, $A$ - произвольная матрица размерности $m \times n$, а B - произвольный вектор размерности $n$, то $A\vec{\xi} + B \sim \mathcal{N}\left( A\vec{\mu}+B, A \Sigma A^\intercal \right)$.

Условное распределение компоненты нормального случайного вектора:
$p_{\xi|\eta=y}(x) = \frac{p_{\xi,\eta}(x,y)}{p_{\eta}(y)}$,
где $p_\eta(y)$ - одномерное нормальное распределение:
$ p_{\eta}(y) = \frac{1}{\sqrt{2\pi}\sqrt{\mathbb{D}_{\eta}}} \exp\left(\frac{-\frac{1}{2}(y-\mathbb{E}_{\eta})^2}{\mathbb{D}_{\eta}}\right) $

#### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi,\eta}(x,y)$.
$p_{\xi,\eta}(x,y)=C\cdot\exp\left(-\frac{1}{2}(7x^2+8xy+13y^2+6x-18y+12)\right)$ <br>
$q(x,y)=7x^2+8xy+13y^2+6x-18y+12=7\left(x^2+\frac{2x(4y+3)}{7}\right)+13y^2-18y+12=\\7\left(x+\frac{4y+3}{y}\right)^2-\frac{(4y+3)^2}{7}+13y^2-18y+12=7\left(x+\frac{4y+3}{y}\right)^2+\frac{75}{7}(y-1)^2=\\7\left(x+1+\frac{4}{7}(y-1)\right)^2+\frac{75}{7}(y-1)^2=7(x+1)^2+8(y-1)(x+1)+13(y-1)^2$
$ \mathbb{E}  \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -1 \\ 1 \end{pmatrix} $
$ \Sigma ^{-1} = \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix}, \quad \mathop{\mathrm{det}}\Sigma^{-1}=75$
$ \Sigma = \frac{1}{75}\begin{pmatrix} 13 & -4 \\ -4 & 7 \end{pmatrix}, \quad \mathop{\mathrm{det}}\Sigma=\frac{1}{75}$
Из ковариационной матрицы получим:
$ \mathbb{D}_{\xi} = \frac{13}{75} \quad \mathbb{D}_{\eta} = \frac{7}{75} $
$ \mathop{\mathrm{cov}}(\xi, \eta) = -\frac{4}{75}  $
$ \rho(\xi, \eta) = \frac{\mathop{\mathrm{cov}}(\xi, \eta)}{\sqrt{\mathbb{D}_{\xi} \cdot \mathbb{D}_{\eta}}} = \frac{-\frac{4}{75}}{\sqrt{\frac{13}{75}\cdot\frac{7}{75}}}= -\frac{4}{\sqrt{91}} $
$ \mathrm{C} = \frac{1}{ (2\pi)^{\frac{n}{2}} \sqrt{\mathop{\mathrm{det}}\Sigma} } = \frac{5\sqrt3}{2\pi} $
$ p_{\xi,\eta}(x,y) = \frac{5\sqrt3}{2\pi} \cdot \exp\left\{ -\frac{1}{2} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ 1 \end{pmatrix} \right]^{\intercal} \cdot \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -1 \\ 1 \end{pmatrix} \right] \right\} $

```python
import numpy as np
import matplotlib.pyplot as plt

def func(X, Y):
    C = 5*np.sqrt(3)/2/np.pi
    Z = C * np.exp(-1/2*(7*X**2+6*X+12+8*X*Y-18*Y+13*Y**2))
    return Z

MU = [-1, 1]
interval = 2

x = np.linspace(MU[0]-interval, MU[0] + interval, 100)
y = np.linspace(MU[1]-interval, MU[1] + interval, 100)

X, Y = np.meshgrid(x, y)

Z = func(X, Y)

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap="winter")

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("График плотности распределения")

plt.show()
```

#### 3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами. Построить график плотности полученного распределения.

$ \Sigma^{-1} = \begin{pmatrix} 7 & 4 \\ 4 & 13 \end{pmatrix} $\
$ \begin{vmatrix} 7-\lambda & 4 \\ 4 & 13-\lambda \end{vmatrix} = (7-\lambda)(13-\lambda) - 16 = 0 $
$ \lambda^{2} - 20\lambda + 75 = 0 \\ \lambda_{1} = 5, \lambda_{2} = 15 $

$ \lambda_{1} = 5: $\
$ \begin{pmatrix} 2 & 4 \\ 4 & 8 \end{pmatrix} \cdot \begin{pmatrix} x \\ y \end{pmatrix} =  \begin{pmatrix} 0 \\ 0 \end{pmatrix} $
$ x+2y=0 \\ y = \alpha \\ x = -2\alpha $
$ \vec{\nu_{1}} = \begin{pmatrix} -2 \\ 1 \end{pmatrix}, \quad ||\vec{\nu_{1}}|| = \sqrt{5} $
Нормируем вектор:
$ \vec\nu_{1}' = \frac{1}{\sqrt5} \cdot \begin{pmatrix} -2 \\ 1 \end{pmatrix}$

$ \lambda_{2} = 15: $\
$ \begin{pmatrix} -8 & 4 \\ 4 & -2 \end{pmatrix} \cdot \begin{pmatrix} x \\ y \end{pmatrix} =  \begin{pmatrix} 0 \\ 0 \end{pmatrix} $
$ 2x-y=0 \\ x = \alpha \\ y = 2\alpha $
$ \vec{\nu_{2}} = \begin{pmatrix} 1 \\ 2 \end{pmatrix}, \quad ||\vec{\nu_{2}}|| = \sqrt{5} $
Нормируем вектор:
$ \vec\nu_{2}' = \frac{1}{\sqrt5} \cdot \begin{pmatrix} 1 \\ 2 \end{pmatrix}$

$\mathrm{Q}^\intercal = \frac{1}{\sqrt5} \cdot \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \Rightarrow \mathrm{Q} = \frac{1}{\sqrt5} \cdot \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} $\
$ \begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix} $\
$ \mathbb{E} \begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \cdot \begin{pmatrix} -1 \\ 1 \end{pmatrix} = \frac{1}{\sqrt{5}} \cdot \begin{pmatrix} 3 \\ 1 \end{pmatrix} $
$ \Sigma_{3} = \mathrm{Q}\Sigma\mathrm{Q}^\intercal = \frac{1}{\sqrt5} \cdot  \frac{1}{\sqrt5} \cdot \frac{1}{75} \cdot \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \cdot \begin{pmatrix} 13 & -4 \\ -4 & 7 \end{pmatrix} \cdot \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} = \begin{pmatrix} \frac{1}{5} & 0 \\ 0 & \frac{1}{15} \end{pmatrix}, \quad \mathop{\mathrm{det}}\Sigma = \frac{1}{75}$

$ \Sigma_{3}^{-1} = \begin{pmatrix} \lambda_{1} & 0 \\ 0 & \lambda_{2} \end{pmatrix} = \begin{pmatrix} 5 & 0 \\ 0 & 15 \end{pmatrix} $\
$ \mathrm{C} =  \frac{1}{ (2\pi)^{\frac{n}{2}} \sqrt{\mathop{\mathrm{det}}\Sigma}} = \frac{5\sqrt3}{2\pi}$

$ p_{\xi_3,\eta_3}(x,y) = \frac{5\sqrt3}{2\pi} \cdot \exp\left\{ -\frac{1}{2} \cdot \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{3}{\sqrt5} \\ \frac{1}{\sqrt5} \end{pmatrix} \right]^{\intercal} \cdot \begin{pmatrix} 5 & 0 \\ 0 & 15 \end{pmatrix} \cdot \left [ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{3}{\sqrt5} \\ \frac{1}{\sqrt5} \end{pmatrix} \right ] \right\} $


```python
def func(X, Y):
    C = 5*np.sqrt(3)/2/np.pi
    Z = C * np.exp(-1/2*(5*X**2-6*np.sqrt(5)*X+12+15*Y**2-6*np.sqrt(5)*Y))
    return Z

MU = [3/np.sqrt(5), 1/np.sqrt(5)]
interval = 2

x = np.linspace(MU[0]-interval, MU[0] + interval, 100)
y = np.linspace(MU[1]-interval, MU[1] + interval, 100)

X, Y = np.meshgrid(x, y)

Z = func(X, Y)

fig = plt.figure(figsize=[10, 10])
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap="winter")

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title("График плотности распределения")

plt.show()
```
