## Основные теоретические положения
Случайный вектор $\vec{\xi} = (\xi_1, \ldots, \xi_n)^{\intercal}: \Omega \rightarrow \mathbb{R}^{n}$ называют нормальным (гауссовым) если существуют вектор независимых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1, ..., \eta_n)^{\intercal}$, матрица $A$ размерности $n \times m$ и вещественный вектор $B = (b_1, \dots, b_m)^{\intercal}$, такие что: $ \vec{\xi} = A\vec{\eta} + B$

Обозначение: $\vec{\xi} \sim \mathcal{N}(\mu, \Sigma)$, где $\mu = \mathbb{E}\vec{\xi} = (\mathbb{E}\xi_1, \dots \mathbb{E}\xi_n)^{\intercal}$ - вектор математических ожиданий компонент; $\Sigma = \{\sigma_{ij}\}^{n}_{ij=1}$, $\sigma_{ij} = \mathop{\mathrm{cov}}(\xi_i,\xi_j)$ - ковариационная матрица.

Плотность нормального вектора имеет вид: $p_{\vec{\xi}}(x_1, ..., x_n) = \frac1{(2\pi)^{\frac{n}2} \sqrt{\det{\Sigma}}} \exp\left(-\frac12 (\vec{x} - \mu)^\intercal \Sigma^{-1} (\vec{x} - \mu) \right)$, где $\vec{x} = (x_1, \dots, x_n)^{\intercal}$.

Свойства: 
1. Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными.
2. Условные и безусловные распределения, полученные из многомерного нормального распределения, также являются нормальными распределениями.
3. Параметры многомерного нормального распределения (средние и ковариационная матрица) полностью описывают его форму.
4. Многомерное нормальное распределение устойчиво относительно линейных преобразований: если $\vec{\xi} \sim \mathcal{N}(\mu, \Sigma)$, $A$ - произвольная матрица неособенная размерности $m \times n$, а  $B$ - произвольный вектор размерности $m \times 1$ то $A\vec{\eta} + B \sim \mathcal{N}(A\mu+B, A \Sigma A^{\intercal})$

### 1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора.
$ q(x,y)=2x^2+4xy+5y^2+8x+2y+11=2(x+(y+2))^2+3(y-1)^2=2((x+3)^2+2(x+3)(y-1)+(y-1)^2)=2(x+3)^2+2\cdot2(x+3)(y-1)+5(y-1)^2 $\
$ \mathbb{E}  \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -3 \\ 1 \end{pmatrix} $ 
$ \Sigma ^{-1} = \begin{pmatrix} 2 & 2 \\ 2 & 5 \end{pmatrix} $
$ \mathop{\mathrm{det}}\Sigma ^{-1} = 6 $
$ \Sigma = \frac16 \cdot \begin{pmatrix} 5 & -2 \\ -2 & 2 \end{pmatrix} $
$ \mathbb{D}_{\xi} = \frac56 \quad \mathbb{D}_{\eta} = \frac13 $ 
$ \mathop{\mathrm{cov}}(\xi, \eta) = -\frac13  $
$ \rho(\xi, \eta) = \frac{cov}{\sqrt{\mathbb{D}_{\xi} \cdot \mathbb{D}_{\eta}}} = -\frac{\sqrt{10}}{5} $
$ C = \frac{1}{ (2\pi)^{\frac{n}{2}} \sqrt{\mathop{\mathrm{det}}\Sigma} } = \frac{\sqrt{6}}{12\pi} $
$ p_{\xi,\eta}(x,y) = \frac{\sqrt{6}}{12\pi} \cdot \exp\left\{ -\frac{1}{2} \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ 1 \end{pmatrix} \right]^{\intercal} \begin{pmatrix} 2 & 2 \\ 2 & 5 \end{pmatrix} \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -3 \\ 1 \end{pmatrix} \right] \right\} $

Построим график плотности распределения $p_{\xi,\eta}$.

```r
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Определение параметров многомерного нормального распределения
mean = np.array([-3, 1])
covariance = np.array([[2, 2], [2, 5]])

# Создание сетки точек для построения графика
x, y = np.meshgrid(np.linspace(-10, 10, 400), np.linspace(-10, 10, 400))
pos = np.dstack((x, y))

# Вычисление значений плотности распределения на сетке точек
rv = multivariate_normal(mean, covariance)
z = rv.pdf(pos)

# Построение графика
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='cividis')
ax.set_xlabel('$x$',fontsize=10')
ax.set_ylabel('$y$',fontsize=10)
ax.set_zlabel('$p_{\\xi,\\eta}(x,y)$',fontsize=10)
ax.set_title('График плотности распределения $p_{\\xi,\\eta}(x,y)$',fontsize=14)
plt.show()
```

### 2. Найти аффинное преобразование, переводящее исходный случайный вектор в стандартный нормальный.
$ q(x,y) = 2(x+(y+2))^2 + 3 (y-1)^2 = \left( \frac{2}{\sqrt{2}} x +  \frac{2}{\sqrt{2}}y + \frac{2}{\sqrt{2}}2 \right)^2 + \left( \frac{3}{\sqrt{3}}y - \frac{3}{\sqrt{3}} \right)^2 $\
$ \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{2}{\sqrt{2}} & \frac{2}{\sqrt{2}} \\ 0 & \frac{3}{\sqrt{3}} \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} \frac{4}{\sqrt{2}} \\ -\frac{3}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} \sqrt{2}\xi + \sqrt{2}\eta + \frac{4\sqrt{3}}{3} \\ \sqrt{3}\eta - \sqrt{3} \end{pmatrix} $
Check:
$ \mathbb{E} \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \begin{pmatrix} \frac{2}{\sqrt{2}} & \frac{2}{\sqrt{2}} \\ 0 & \frac{3}{\sqrt{3}} \end{pmatrix} \mathbb{E} \begin{pmatrix} \xi \\ \eta \end{pmatrix} + \begin{pmatrix} \frac{4}{\sqrt{2}} \\ -\frac{3}{\sqrt{3}} \end{pmatrix} = \begin{pmatrix} \frac{2}{\sqrt{2}} & \frac{2}{\sqrt{2}} \\ 0 & \frac{3}{\sqrt{3}} \end{pmatrix} \begin{pmatrix} -3 \\ 1 \end{pmatrix} +  \begin{pmatrix} \frac{4}{\sqrt{2}} \\ -\frac{3}{\sqrt{3}} \end{pmatrix} =  \begin{pmatrix} 0 \\ 0 \end{pmatrix} $\

### 3. Найти ортогональное преобразование, переводящее соответствующий центрированный случайный вектор в вектор с независимыми компонентами.
$ \Sigma^{-1} = \begin{pmatrix} 2 & 2 \\ 2 & 5 \end{pmatrix} $\
$ \begin{vmatrix} 2-\lambda & 2 \\ 2 & 5-\lambda \end{vmatrix} = (2-\lambda)(5-\lambda) - 4 = 0 $
$ \lambda^{2} - 7\lambda + 6 = 0 \quad \lambda_{1} = 1, \lambda_{2} = 6 $

$ \lambda_{1} = 1: $\
$ \begin{pmatrix} 1 & 2 \\ 2 & 4 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} =  \begin{pmatrix} 0 \\ 0 \end{pmatrix} $\
$ x+2y=0 \quad y = \alpha, x = -2\alpha \quad \vec{\nu_{1}} = \begin{pmatrix} -2 \\ 1 \end{pmatrix} \quad ||\vec{\nu_{1}}|| = \sqrt{5} $ 
$ \nu_{1}' = \begin{pmatrix} -\frac{2}{\sqrt{5}} \\ \frac{1}{\sqrt{5}} \end{pmatrix}$

$ \lambda_{2} = 6: $\
$ \begin{pmatrix} -4 & 2 \\ 2 & -1 \end{pmatrix} \begin{pmatrix} x \\ y \end{pmatrix} =  \begin{pmatrix} 0 \\ 0 \end{pmatrix} $\
$ 2x-y=0 \quad x = \alpha, y = 2\alpha \quad \vec{\nu_{2}} = \begin{pmatrix} 1 \\ 2 \end{pmatrix} \quad ||\vec{\nu_{2}}|| = \sqrt{5} $ 
$ \nu_{2}' = \begin{pmatrix} \frac{1}{\sqrt{5}} \\ \frac{2}{\sqrt{5}} \end{pmatrix} $

Check:

$ \begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \begin{pmatrix} \xi \\ \eta \end{pmatrix} $\
$ \mathbb{E} \begin{pmatrix} \xi_{3} \\ \eta_{3} \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \begin{pmatrix} -3 \\ 1 \end{pmatrix} = \frac{1}{\sqrt{5}} \begin{pmatrix} 7 \\ -1 \end{pmatrix} $
$ \Sigma_{3} = \frac{1}{\sqrt{5}} \cdot \frac {1}{\sqrt{5}} \cdot \frac{1}{6} \cdot \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} \begin{pmatrix} 5 & -2 \\ -2 & 2 \end{pmatrix} \begin{pmatrix} -2 & 1 \\ 1 & 2 \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & \frac{1}{6} \end{pmatrix} = \begin{pmatrix} \frac{1}{\lambda_{1}} & 0 \\ 0 & \frac{1}{\lambda_{2}} \end{pmatrix} $
$ \Sigma^{-1}_{3} = \begin{pmatrix} \lambda_{1} & 0 \\ 0 & \lambda_{2} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 6 \end{pmatrix} $
$ p_{\xi_3, \eta_3}(x,y) = \frac{\sqrt{6}}{12\pi} \cdot \exp\left\{ -\frac{1}{2} \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{7}{\sqrt{5}} \\ -\frac{1}{\sqrt{5}} \end{pmatrix} \right]^{\intercal} \begin{pmatrix} 1 & 0 \\ 0 & 6 \end{pmatrix} \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} \frac{7}{\sqrt{5}} \\ -\frac{1}{\sqrt{5}} \end{pmatrix} \right] \right\} $

Построим график плотности полученного распределения. 

```r
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Определение параметров многомерного нормального распределения
mean = np.array([7/np.sqrt(5), -1/np.sqrt(5)])
covariance = np.array([[1, 0], [0, 6]])

# Создание сетки точек для построения графика
x, y = np.meshgrid(np.linspace(-10, 10, 400), np.linspace(-10, 10, 400))
pos = np.dstack((x, y))

# Вычисление значений плотности распределения на сетке точек
rv = multivariate_normal(mean, covariance)
z = rv.pdf(pos)

# Построение трехмерного графика
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='cividis') 
ax.set_xlabel('$x$', fontsize=10)
ax.set_ylabel('$y$', fontsize=10)
ax.set_zlabel('$p_{\\xi_3,\\eta_3}(x,y)$', fontsize=10)
ax.set_title('График плотности распределения $p_{\\xi_3,\\eta_3}(x,y)$', fontsize=14)
plt.show()
```

### 4. Вычислить характеристики совместного распределения случайного вектора $(2 \xi - 2 \eta, -5 \xi + 2 \eta)^{\intercal} $ и записать его плотность.
$ \mathbb{E} \begin{pmatrix} \xi_4 \\ \eta_4 \end{pmatrix} = \begin{pmatrix} 2 & -2 \\ -5 & 2 \end{pmatrix} \begin{pmatrix} -3 \\ 1 \end{pmatrix} = \begin{pmatrix} -8 \\ 17 \end{pmatrix} $\
$ \Sigma_4 = \begin{pmatrix} 2 & -2 \\ -5 & 2 \end{pmatrix} \begin{pmatrix} \frac{5}{6} & -\frac{2}{6} \\ -\frac{2}{6} & \frac{2}{6} \end{pmatrix} \begin{pmatrix} 2 & -5 \\ -2 & 2 \end{pmatrix} = \begin{pmatrix} \frac{22}{3} & -\frac{43}{6} \\ -\frac{43}{3} & \frac{173}{6} \end{pmatrix}$\
$ \mathop{\mathrm{det}} \Sigma_4 = 6 $
$ \Sigma^{-1}_4 = \begin{pmatrix} \frac{173}{36} & \frac{43}{18} \\ \frac{43}{18} & \frac{11}{9} \end{pmatrix} $
$ \mathbb{D}(\xi_4) = \frac{22}{3} \quad \mathbb{D}(\eta_4) = \frac{173}{6} $
$ \mathop{\mathrm{cov}}(\xi_4, \eta_4) = -\frac{43}{3} $
$ \rho(\xi_4, \eta_4) = -\frac{43}{\sqrt{1903}} $
$ C = \frac{1}{ (2\pi)^{\frac{n}{2}} \sqrt{\mathop{\mathrm{det}}\Sigma_4} } = \frac{\sqrt{6}}{12\pi} $
$ p_{\xi_4, \eta_4}(x,y) = \frac{\sqrt{6}}{12\pi} \cdot \exp\left\{ -\frac{1}{2} \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -8 \\ 17 \end{pmatrix} \right]^{\intercal} \begin{pmatrix} \frac{173}{36} & \frac{43}{18} \\ \frac{43}{18} & \frac{11}{9} \end{pmatrix} \left[ \begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -8 \\ 17 \end{pmatrix} \right] \right\} $


Построим график плотности полученного распределения.

```r
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Определение параметров многомерного нормального распределения
mean = np.array([-8, 17])
covariance = np.array([[173/36, 43/18], [43/18, 11/9]])

# Создание сетки точек для построения графика
x, y = np.meshgrid(np.linspace(-20, 10, 100), np.linspace(0, 30, 100))
pos = np.dstack((x, y))

# Вычисление значений плотности распределения на сетке точек
rv = multivariate_normal(mean, covariance)
z = rv.pdf(pos)


fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='cividis')  
ax.set_xlabel('$x$', fontsize=10)  
ax.set_ylabel('$y$', fontsize=10)  
ax.set_zlabel('$p_{\\xi_4,\\eta_4}(x,y)$', fontsize=10)  
ax.set_title('График плотности распределения $p_{\\xi_4,\\eta_4}(x,y)$', fontsize=16)  
plt.show()
```

### 5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E} ( \xi | \eta )$ и $\mathbb{D} ( \xi | \eta )$
