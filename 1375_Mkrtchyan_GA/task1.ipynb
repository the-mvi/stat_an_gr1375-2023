## Основные теоретические положения
- Случайный вектор $\vec\xi=(\xi_1,\ \xi_2,\ ...,\ \xi_n)^\intercal: \Omega \rightarrow \mathbb{R}^n$ называют нормальным (гауссовым), если существует вектор независимых стандартных нормальных случайных величин $\vec{\eta} = (\eta_1,\ ...,\ \eta_n)^\intercal$, матрица $А$ размерности $n \times m$ и вещественный вектор $B = (b_1,\ ...,\ b_n)^\intercal$, такие что: $\vec\xi=А\vec\eta+B$


- Многомерное нормальное распределение
    1. Линейные комбинации случайных переменных, принадлежащих многомерному нормальному распределению, также являются нормально распределенными
    2. Условные и безусловные распределения, полученные из многомерного распределения, также являются нормальными распределениями
    3. Параметры многомерного нормального распределения (средние и ковариацционная матрица) полностью описывает его форму
    4. Многомерное нормальное распределение устойчиво относительно линейных преобразований, если $\vec\xi  \sim \mathcal{N}(\vec{\mu},\ \Sigma),\ A$ - произвольная матрица неособенная размерности $n \times m$, а $B$ - произвольный вектор размерности $m \times 1$ то: $A\vec{\xi} + B \sim \mathcal{N}(A\vec{\mu} + B,\ A \Sigma A^\intercal)$


- Ковариационная матрица - $\Sigma = \frac{1}{12}\begin{pmatrix} \mathbb{D}_\xi & \mathbb{cov}(\xi, \eta) \\ \mathbb{cov}(\eta, \xi) & \mathbb{D}_\eta\end{pmatrix}$


- Плотность нормального вектора - $p_{\vec{\xi}}(\vec{x}) = \frac{1}{(2\pi)^{\frac{n}{2}}\sqrt{det\Sigma}} \cdot \exp \left\{-\frac{1}{2} \Biggl[\vec{x}-\mu\Biggr]^{\intercal}\Sigma^{-1} \Biggl[\vec{x}-\mu \Biggr] \right\}$

## Выполнение работы
1. Вычислить вектор математических ожиданий и ковариационные характеристики данного случайного вектора. Построить график плотности распределения $p_{\xi,\eta}$
$ 3x^2 - 3xy + 7y^2 + 15x + 5y + 25 = 3(x^2 - x(y - 5)) + 7y^2 + 5y + 25 = 3 \left(x^2 - x(y - 5) + \frac{(y - 5)^2}{4} \right) - 3\frac{(y - 5)^2}{4} + 7y^2 + 5y + 25 = 3 \left(x - \frac{y - 5}{2} \right)^2 +\frac{25y^2 + 50y + 25}{4} = $ <br>
$ = 3 \left(x - \frac{y - 5}{2} \right)^2 + \frac{25(y + 1)^2}{4} = 3 \left((x + 3) - \frac{1}{2}(y + 1)\right)^2 +\frac{25}{4}(y + 1)^2 = 3(x + 3)^2 - 3(x + 3)(y + 1) + 7(y + 1)^2 $ <br>
$ E {\xi \choose \eta} = {-3 \choose -1} $ <br>
$ \Sigma^{-1} = \begin{pmatrix} 3 & -1.5\\ -1.5 & 7\end{pmatrix},\ \Sigma = \frac{4}{75}\begin{pmatrix} 7 & 1.5 \\ 1.5 & 3\end{pmatrix} $ <br>
$ \mathbb{\mathrm{cov}}(\xi, \eta) = \frac{2}{25},\ \mathbb{D}_\xi = \frac{28}{75},\ \mathbb{D}_\eta = \frac{4}{25} $ <br>
$ \rho(\xi,\eta) = \frac{\mathbb{\mathrm{cov}}(\xi, \eta)}{\sqrt{\mathbb{D}_\xi} \sqrt{\mathbb{D}_\eta}} = \frac{\frac{2}{25}}{\sqrt{\frac{28}{75}} \sqrt{\frac{4}{25}}} = \frac{\sqrt{21}}{14} $ <br>
$ C = \frac{1}{2\pi\sqrt{\mathrm{det\Sigma}}} = \frac{1}{2\pi\sqrt{\frac{4}{75}}} = \frac{5 \sqrt{3}}{4 \pi} $ <br>
$ p_{\xi,\eta}(x, y) = \frac{5 \sqrt{3}}{4 \pi} \cdot \exp \left\{-\frac{1}{2} \Biggl[\begin{pmatrix} x \\ y\end{pmatrix} - \begin{pmatrix} 3 \\ 1\end{pmatrix} \Biggr]^{\intercal}\begin{pmatrix} 3 & -1.5\\ -1.5 & 7\end{pmatrix} \Biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} 3 \\ 1\end{pmatrix} \Biggr]\right\} $ <br>

```python
import numpy as numpy
import matplotlib.pyplot as plot

coefficient = (5 * numpy.sqrt(3)) / (4 * numpy.pi)

def f(x, y):
    exponent = -0.5 * (3 * x ** 2 - 3 * x * y + 7 * y ** 2 + 15 * x + 5 * y + 25)
    return coefficient * numpy.exp(exponent)

X, Y = numpy.meshgrid(numpy.linspace(-6, 0, 1000), numpy.linspace(-4, 2, 1000))
Z = f(X, Y)

figure = plot.figure(figsize=(10, 10))

axes = figure.add_subplot(111, projection='3d')
axes.set_facecolor('white')
surface = axes.plot_surface(X, Y, Z, cmap='cool')

axes.set_xlabel('X')
axes.set_ylabel('Y')

plot.show()
```

\vec{\nu_{1}} = \begin{pmatrix} 1 \\ -3 \end{pmatrix}, \ ||\vec{\nu_{1}}|| = \sqrt{10}, \ \vec{\nu_{1}} = \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 1 \\ -3 \end{pmatrix} $ <br>
$ \mathrm{Q}^\intercal = \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} \Rightarrow \mathrm{Q} = \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} $ <br>
$ \begin{pmatrix} \xi_{st} \\ \eta_{st} \end{pmatrix} = \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix} $ <br>
$ \mathbb{E}\begin{pmatrix} \xi_{st} \\ \eta_{st}\end{pmatrix} = \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} \cdot \begin{pmatrix} -3 \\ -1 \end{pmatrix} = \begin{pmatrix} -\sqrt{10} \\ 0 \end{pmatrix} $ <br>
$ \Sigma_{3} = \frac{4}{75} \cdot \frac{1}{\sqrt{10}} \cdot \begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} \cdot \begin{pmatrix} 7 & 1.5 \\ 1.5 & 3 \end{pmatrix} \cdot \begin{pmatrix} 3 & 1 \\ 1 & -3 \end{pmatrix} = \begin{pmatrix} \frac{2\sqrt{10}}{5} & 0 \\ 0 & \frac{2\sqrt{10}}{15} \end{pmatrix} $ <br>

$ p_{\xi_3,\eta_3}(x, y) = \frac{\sqrt{30}}{8 \pi} \cdot \exp \left\{-\frac{1}{2} \Biggl[\begin{pmatrix} x \\ y \end{pmatrix} - \begin{pmatrix} -\sqrt{10} \\ 0 \end{pmatrix} \Biggr]^{\intercal}\begin{pmatrix} 2.5 & 0 \\ 0 & 7.5 \end{pmatrix} \Biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} -\sqrt{10} \\ 0 \end{pmatrix} \Biggr]\right\} $ <br>


```python
coefficient = (numpy.sqrt(30)) / (8 * numpy.pi)

def f(x, y):
    exponent = -0.5 * (2.5 * x ** 2 + 7.5 * y ** 2 + 5 * numpy.sqrt(10) * x + 25)
    return coefficient * numpy.exp(exponent)

X, Y = numpy.meshgrid(numpy.linspace(-6, 0, 1000), numpy.linspace(-3, 3, 1000))
Z = f(X, Y)

figure = plot.figure(figsize=(10, 10))

axes = figure.add_subplot(111, projection='3d')
axes.set_facecolor('white')
surface = axes.plot_surface(X, Y, Z, cmap='cool')

axes.set_xlabel('X')
axes.set_ylabel('Y')

plot.show()
```

4. Вычислить характеристики совместного распределения случайного вектора $(-5\xi + 4\eta,\ 2\xi + \eta)^{\intercal}$  записать его плотность. Построить график плотности полученного распределения
$ \mathbb{E}\begin{pmatrix} \xi_{4} \\ \eta_{4} \end{pmatrix} = \begin{pmatrix} -5 & 4 \\ 2 & 1 \end{pmatrix} \cdot \begin{pmatrix} \xi \\ \eta \end{pmatrix} = \begin{pmatrix} -5 & 4 \\ 2 & 1 \end{pmatrix} \cdot \begin{pmatrix} -3 \\ -1 \end{pmatrix} = \begin{pmatrix} 11 \\ -7 \end{pmatrix} $ <br>
$ \Sigma_{4} = \begin{pmatrix} 2.5 & 0 \\ 0 & 7.5 \end{pmatrix} = \frac{4}{75} \cdot \begin{pmatrix} -5 & 4 \\ 2 & 1 \end{pmatrix} \cdot \begin{pmatrix} 7 & 1.5 \\ 1.5 & 3 \end{pmatrix} \cdot \begin{pmatrix} -5 & 2 \\ 4 & 1 \end{pmatrix} = \frac{4}{75} \cdot \begin{pmatrix} 7 & 1.5 \\ 1.5 & 3 \end{pmatrix} = \frac{4}{75} \cdot \begin{pmatrix} 163 & -53.5 \\ -53.5 & 37 \end{pmatrix} $ <br>
$ \Sigma^{-1}_{4} = \frac{4}{75} \cdot \frac{4}{12675} \cdot \begin{pmatrix} 37 & 53.5 \\ 53.5 & 163\end{pmatrix} $ <br>
$ \mathbb{\mathrm{cov}}(\xi, \eta) = \frac{214}{75},\ \mathbb{D}_\xi = \frac{652}{75},\ \mathbb{D}_\eta = \frac{148}{75} $ <br>
$ \rho(\xi,\eta) = \frac{\mathbb{\mathrm{cov}}(\xi, \eta)}{\sqrt{\mathbb{D}_\xi} \sqrt{\mathbb{D}_\eta}} = \frac{\frac{214}{75}}{\sqrt{\frac{652}{75}} \sqrt{\frac{148}{75}}} = \frac{107\sqrt{6031}}{12062} $ <br>
$ \mathrm{det}\Sigma_4 = \frac{676}{75} $ <br>
$ C = \frac{1}{2\pi\sqrt{det\Sigma}} = \frac{1}{2\pi\sqrt{\frac{676}{75}}} = \frac{5 \sqrt{3}}{52 \pi} $ <br>
$ p_{\xi,\eta}(x, y) = \frac{5 \sqrt{3}}{52 \pi} \cdot \exp \left\{-\frac{1}{2} \Biggl[\begin{pmatrix} x \\ y\end{pmatrix} - \begin{pmatrix} 11 \\ -7 \end{pmatrix} \Biggr]^{\intercal} \frac{4}{75} \cdot \frac{4}{12675} \cdot \begin{pmatrix} 37 & 53.5 \\ 53.5 & 163\end{pmatrix} \Biggl[\begin{pmatrix} x \\ y\end{pmatrix}-\begin{pmatrix} 11 \\ -7 \end{pmatrix} \Biggr]\right\} $ <br>

```python
coefficient = (5 * numpy.sqrt(3)) / (52 * numpy.pi)

def f(x, y):
    exponent = -0.5 * (16 / 950625) * (37 * x ** 2 + 107 * x * y + 163 * y ** 2 - 65 * x + 1105 * y + 4225)
    return coefficient * numpy.exp(exponent)

X, Y = numpy.meshgrid(numpy.linspace(-200, 200, 1000), numpy.linspace(-200, 200, 1000))
Z = f(X, Y)

figure = plot.figure(figsize=(10, 10))

axes = figure.add_subplot(111, projection='3d')
axes.set_facecolor('white')
surface = axes.plot_surface(X, Y, Z, cmap='cool')

axes.set_xlabel('X')
axes.set_ylabel('Y')

plot.show()
```

5. Найти условное распределение $\xi$ при условии $\eta$. Вычислить $\mathbb{E}(\xi | \eta)$ и $\mathbb{D}(\xi | \eta) $
$ p_{\xi|\eta=y} = \frac{p_{\xi,\eta}\left(x,y\right)}{p_{\eta}\left(y\right)} $ <br>
$ p_{\eta}(y) = \frac{1}{\sqrt{2\pi} \sqrt{\mathbb{D}_{\eta}}} \exp \left\{\frac{-\frac{1}{2}(y - \mathbb{E}_{\eta})^2}{\mathbb{D}_{\eta}}\right\} $ <br>
$ p_{\eta}(y) = \frac{1}{\sqrt{2\pi} \sqrt{\frac{4}{25}}} \exp \left\{\frac{-\frac{1}{2}(y - (-1))^2}{\frac{4}{25}} \right\} = \frac{5\sqrt{2\pi}}{4\pi} \exp \left\{\frac{-\frac{1}{2}(y + 1)^2}{\frac{4}{25}} \right\} $ <br>
$ p_{\xi|\eta=y} = \frac{\frac{5 \sqrt{3}}{4 \pi} \exp \left\{-\frac{1}{2}(3x^2 - 3xy + 7y^2 + 15x + 5y + 25) \right\}}{\frac{5\sqrt{2\pi}}{4\pi} \exp \left\{-\frac{1}{2}\frac{(y + 1)^2}{\frac{4}{25}} \right\}} = \frac{\sqrt{6\pi}}{2\pi} \exp \left\{-\frac{1}{2} \left(3x^2 - 3xy + 7y^2 + 15x + 5y + 25 - \frac{(y + 1)^2}{\frac{4}{25}} \right) \right\} $ <br>
$ 3(x^2 - x(y - 5)) + 7y^2 + 5y + 25 - \frac{(y + 1)^2}{\frac{4}{25}} = 3 \left(x^2 - x(y - 5) + \frac{(y - 5)^2}{4} \right) - 3\frac{(y - 5)^2}{4} + 7y^2 + 5y + 25 - \frac{(y + 1)^2}{\frac{4}{25}} = 3 \left(x - \frac{y - 5}{2} \right)^2 = 3 \left(x - \frac{y - 5}{2} \right)^2 + \frac{25y^2 + 50y + 25}{4} - \frac{(y + 1)^2}{\frac{4}{25}} = 3 \left(x - \frac{y - 5}{2} \right)^2 $ <br>
$ p_{\xi|\eta=y} = \frac{\sqrt{6\pi}}{2\pi} \exp \left\{-\frac{1}{2} \left(3 \left(x - \frac{y - 5}{2} \right)^2 \right) \right\} $ <br>
$ \mathbb{E}(\xi | \eta) = \left(\frac{y}{2} - \frac{5}{2} \right),\  \mathbb{D}(\xi | \eta) = \frac{1}{3} $ <br>
